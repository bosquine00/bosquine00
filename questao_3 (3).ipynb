## Questão 3a)

**Enunciado:** Autovetores: Seja  $A(\omega) =\begin{pmatrix}0 & 1 \\-\omega^2 & 0\end{pmatrix}$.
Faça um gráfico dos autovalores para $\omega$ variando de 0 a 1, e também dos autovetores. O que acontece quando $\omega \to 0^+$?


### 1. Cálculo dos Autovalores ($\lambda$)

Os autovalores são determinados pelas raízes do polinômio característico, dado por $\det(A(\omega) - \lambda I) = 0$:

$$\det \begin{bmatrix} -\lambda & 1 \\ -\omega^2 & -\lambda \end{bmatrix} = 0$$

$$\lambda^2 - (1)(-\omega^2) = 0 \implies \lambda^2 + \omega^2 = 0$$

Como o enunciado define $\omega$ variando de $0$ a $1$, as raízes são puramente imaginárias:
$$\lambda_1 = i\omega, \quad \lambda_2 = -i\omega$$

**Comportamento no Gráfico:** No plano complexo (onde o eixo horizontal representa a parte Real e o eixo vertical a parte Imaginária), conforme $\omega$ cresce de $0$ para $1$, os autovalores partem juntos da origem $(0,0)$ e se movem em direções opostas ao longo do eixo imaginário, terminando em $i$ e $-i$.

---

### 2. Cálculo dos Autovetores ($v$)

Para encontrar os autovetores associados a cada autovalor, resolvemos o sistema linear homogêneo $(A(\omega) - \lambda I)v = 0$.

#### Para $\lambda_1 = i\omega$:
$$\begin{bmatrix} -i\omega & 1 \\ -\omega^2 & -i\omega \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$$

Multiplicando a primeira linha:
$$-i\omega x_1 + x_2 = 0 \implies x_2 = i\omega x_1$$

Adotando a parametrização livre $x_1 = 1$, obtemos o primeiro autovetor:
$$v_1 = \begin{bmatrix} 1 \\ i\omega \end{bmatrix}$$

#### Para $\lambda_2 = -i\omega$:
$$\begin{bmatrix} i\omega & 1 \\ -\omega^2 & i\omega \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$$

Multiplicando a primeira linha:
$$i\omega x_1 + x_2 = 0 \implies x_2 = -i\omega x_1$$

Adotando a parametrização livre $x_1 = 1$, obtemos o segundo autovetor:
$$v_2 = \begin{bmatrix} 1 \\ -i\omega \end{bmatrix}$$

---

### 3. Análise do Limite quando $\omega \to 0^+$

Ao avaliar o comportamento do sistema quando o parâmetro $\omega$ tende a zero por valores positivos, observamos as seguintes consequências algébricas:

* **Colapso dos Autovalores:**
  $$\lim_{\omega \to 0^+} \lambda_1 = \lim_{\omega \to 0^+} \lambda_2 = 0$$
  Ambos os autovalores distintos convergem para o mesmo ponto (a origem). A matriz limite $A(0) = \begin{bmatrix} 0 & 1 \\ 0 & 0 \end{bmatrix}$ passa a ter um único autovalor $\lambda = 0$ com **multiplicidade algébrica igual a 2**.

* **Colapso dos Autovetores:**
  $$\lim_{\omega \to 0^+} v_1 = \lim_{\omega \to 0^+} v_2 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$$
  À medida que $\omega$ decresce em direção a zero, as duas direções próprias (que an$tes eram linearmente independentes no espaço complexo) se alinham e colapsam no mesmo vetor. Portanto, a matriz limite possui apenas **multiplicidade geométrica igual a 1**.

---
### Gráfico dos autovalores
Foram construídos 3 gráficos acerca do comportamento dos autovalores em função de $\omega$:
- Plano complexo dos autovalores;
- Módulo do autovalor por $\omega$;
- Parte imaginária do autovalor por $\omega$.


```python
import numpy as np 

ws = np.linspace(0,1, 200)
lista_lambda_1 = []
lista_lambda_2 = []

for w in ws:
    Aw = np.array([[0, 1], [-w**2, 0]])
    eigvals = np.linalg.eigvals(Aw)

    # ordena pela parte imaginária
    eigvals = sorted(eigvals, key=lambda z: z.imag)

    lista_lambda_1.append(eigvals[0])
    lista_lambda_2.append(eigvals[1])
```

```python
# PLOT

import matplotlib.pyplot as plt

lista_lambda_1 = np.array(lista_lambda_1)
lista_lambda_2 = np.array(lista_lambda_2)

plt.figure()
plt.plot(lista_lambda_1.real, lista_lambda_1.imag, label="λ = +iω")
plt.plot(lista_lambda_2.real, lista_lambda_2.imag, label="λ = -iω")

plt.xlabel("Parte real")
plt.ylabel("Parte imaginária")
plt.title("Autovalores de A(ω) no plano complexo")
plt.legend()
plt.grid()
plt.show()

plt.figure()
plt.plot(ws, ws, label="|λ| = ω")
plt.xlabel("ω")
plt.ylabel("|λ|")
plt.title("Módulo dos autovalores")
plt.grid()
plt.legend()
plt.show()

plt.figure()
plt.plot(ws, lista_lambda_1.imag, label="Im(λ+) = ω")
plt.plot(ws, lista_lambda_2.imag, label="Im(λ-) = -ω")

plt.xlabel("ω")
plt.ylabel("Parte imaginária")
plt.title("Autovalores vs ω")
plt.grid()
plt.legend()
plt.show()
```

### Conclusões da Análise Gráfica


#### Gráfico 1: Autovalores de $A(\omega)$ no Plano Complexo
**O que observamos:** Duas retas sobrepostas perfeitamente ao eixo vertical (eixo imaginário), partindo da origem e divergindo em direções opostas (+ e -).

**Conclusão:** como mostrado matematicamente, os testes empíricos provam que $\lambda_i$ é puramente imaginário.

#### Gráfico 2: Módulo dos autovalores
**O que observamos:** Uma reta diagonal linear perfeita partindo da origem, correspondendo à função identidade ($y = x$).

**Conclusão:** O módulo representa a norma, ou seja, a "distância" do autovalor até a origem no plano complexo. Matematicamente, isso ocorre porque:

$$|\lambda_{1,2}| = \sqrt{0^2 + (\pm \omega)^2} = \omega$$

Isso conclui que a norma espectral cresce de forma idêntica e linear para ambas as raízes.

#### Gráfico 3: Autovalores vs $\omega$ (Parte Imaginária)
**O que observamos:** Um formato de "X", com duas retas concorrentes na origem (uma crescendo linearmente e outra decrescendo linearmente à mesma taxa).

**Conclusão:** Este gráfico evidencia o cumprimento do **Teorema das Raízes Complexas Conjugadas**. Como a matriz $A(\omega)$ possui apenas entradas reais (0, 1 e $-\omega^2$), as raízes imaginárias são obrigadas a aparecer em pares espelhados. O gráfico mapeia essa complementaridade exata:

$$\text{Im}(\lambda_+) = \omega \quad \text{e} \quad \text{Im}(\lambda_-) = -\omega$$

#### Conclusão Geral
Ao cruzarmos a informação dos três gráficos, a conclusão mais importante é o que acontece na extremidade esquerda (origem). 

Quando fazemos o limite $\omega \to 0$, o módulo zera, a conjugação desaparece e as partes imaginárias colapsam simultaneamente no ponto $(0,0)$.

### Gráfico dos autovetores

Sendo o autovetor $v=(x_1,x_2)$, iremos mostrar o comportamento de $x_1$ no plano real e de $x_2$ no plano complexo.

```python
import matplotlib.pyplot as plt

ws = np.linspace(1e-6, 1, 500)

v1_x1, v2_x1 = [], []
v1_x2, v2_x2 = [], []

for w in ws:

    Aw = np.array([
        [0, 1],
        [-w**2, 0]
    ], dtype=complex)

    eigvals, eigvecs = np.linalg.eig(Aw)

    # Normaliza para que a primeira componente seja 1
    v1 = eigvecs[:, 0] / eigvecs[0, 0]
    v2 = eigvecs[:, 1] / eigvecs[0, 1]

    v1_x1.append(v1[0])
    v1_x2.append(v1[1])

    v2_x1.append(v2[0])
    v2_x2.append(v2[1])

v1_x1 = np.array(v1_x1)
v2_x1 = np.array(v2_x1)

v1_x2 = np.array(v1_x2)
v2_x2 = np.array(v2_x2)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico da esquerda
ax[0].plot(ws, v1_x1.real, label=r"$x_1$ de $v_1$")
ax[0].plot(ws, v2_x1.real, "--", label=r"$x_1$ de $v_2$")

ax[0].set_title(r"Componente $x_1$")
ax[0].set_xlabel(r"$\omega$")
ax[0].set_ylabel(r"$x_1$")
ax[0].grid(True, linestyle="--")
ax[0].legend()

# Gráfico da direita
ax[1].plot(ws, np.imag(v1_x2),
           label=r"$\mathrm{Im}(x_2(v_1))$")

ax[1].plot(ws, np.imag(v2_x2),
           "--",
           label=r"$\mathrm{Im}(x_2(v_2))$")

ax[1].set_title(r"Parte imaginária de $x_2$")
ax[1].set_xlabel(r"$\omega$")
ax[1].set_ylabel(r"$\mathrm{Im}(x_2)$")
ax[1].grid(True, linestyle="--")
ax[1].legend()

plt.tight_layout()
fig.suptitle(
    "Comportamento dos Autovetores em Função de $\\omega$",
    fontsize=14, 
    y = 1.05
)

plt.show()
```

Como esperado da demonstração teórica no exercício, $x_1 = 1$, para qualquer um dos dois autovetores e $x_2$ varia no plano complexo entre $-1$ e 1, o que reforça $x_2 =  \pm i\omega$.

A decomposição do comportamento em dois gráficos lado a lado em função de $\omega$. À esquerda, a parte real das componentes é representada por uma linha reta sobre o zero. À direita, a parte imaginária mostra duas retas simétricas divergindo da origem, atingindo $+1$ e $-1$.


O gráfico da esquerda atesta que $\text{Re}(v_y) = 0$ globalmente, blindando a prova de que não há contribuição real cruzada. O gráfico da direita expõe a taxa de variação constante (linear) das partes imaginárias com o parâmetro $\omega$. A simetria perfeita em "V deitado" no gráfico da direita é a assinatura visual de autovetores complexos conjugados, que sempre se separam simetricamente no eixo imaginário a partir de um ponto de ramificação.

## Questão 3b)

**Enunciado:** Condicionamento. Calcule (no computador, mas verifique analiticamente!) uma
decomposição SVD de A(ω), e deduza o número de condicionamento de A. Qual o condicionamento da matriz de autovetores de A(ω), em função de ω?

### Resolução do Item b) Condicionamento

Dada a matriz paramétrica:
$$A(\omega) = \begin{bmatrix} 0 & 1 \\ -\omega^2 & 0 \end{bmatrix}$$

Para simplificar a álgebra e manter a consistência com o desenvolvimento analítico, adota-se a mudança de variáveis:
$$\omega = i \cdot b \implies -\omega^2 = b^2$$

Dessa forma, trabalharemos com a matriz na forma:
$$A = \begin{bmatrix} 0 & 1 \\ b^2 & 0 \end{bmatrix}$$


#### 1. Decomposição SVD e Número de Condicionamento de $A(\omega)$

Para deduzir os valores singulares de $A$, calculamos a matriz $A^*A$ (onde $A^*$ representa a transposta conjugada):

$$A^* = \begin{bmatrix} 0 & \overline{b^2} \\ 1 & 0 \end{bmatrix}$$

Como $-\omega^2 = b^2$ é um valor real no contexto da mudança de variáveis adotada, temos $\overline{b^2} = b^2$. Portanto:

$$A^*A = \begin{bmatrix} 0 & b^2 \\ 1 & 0 \end{bmatrix} \begin{bmatrix} 0 & 1 \\ b^2 & 0 \end{bmatrix} = \begin{bmatrix} b^4 & 0 \\ 0 & 1 \end{bmatrix}$$

Os autovalores de $A^*A$ são obtidos diretamente pela equação característica:
$$\det(A^*A - \lambda I) = 0 \implies \det \begin{bmatrix} b^4 - \lambda & 0 \\ 0 & 1 - \lambda \end{bmatrix} = 0$$

$$\implies \lambda_1 = b^4 \quad \text{ou} \quad \lambda_2 = 1$$

Como o número de condicionamento na norma-$2$ é a razão entre o maior e o menor valor singular ($\sigma = \sqrt{\lambda}$), assumindo $b^4 \le 1$:
$$\kappa(A_\omega) = \frac{\sigma_{\max}}{\sigma_{\min}} = \frac{\sqrt{1}}{\sqrt{b^4}} = \frac{1}{b^2}$$

Retornando para a variável original $\omega$ (onde $b^2 = -\omega^2$, tomando o módulo para garantir a positividade da norma):
$$\kappa(A(\omega)) = \frac{1}{\omega^2}$$


#### 2. Condicionamento da Matriz de Autovetores $M$

Deseja-se encontrar a matriz de autovetores $M$ de $A = \begin{bmatrix} 0 & 1 \\ b^2 & 0 \end{bmatrix}$. O polinômio característico de $A$ é:
$$\det(A - \lambda I) = \det \begin{bmatrix} -\lambda & 1 \\ b^2 & -\lambda \end{bmatrix} = \lambda^2 - b^2 = 0 \implies \lambda = b \quad \text{ou} \quad \lambda = -b$$

##### Determinação dos Autovetores:
* **Para $\lambda = b$:**
  $$\begin{bmatrix} -b & 1 \\ b^2 & -b \end{bmatrix} \to \begin{bmatrix} -b & 1 \\ 0 & 0 \end{bmatrix} \implies v_1 = \begin{bmatrix} -1 \\ b \end{bmatrix}$$

* **Para $\lambda = -b$:**
  $$\begin{bmatrix} b & 1 \\ b^2 & b \end{bmatrix} \to \begin{bmatrix} b & 1 \\ 0 & 0 \end{bmatrix} \implies v_2 = \begin{bmatrix} 1 \\ b \end{bmatrix}$$

Montando a matriz de autovetores $M$:
$$M = \begin{bmatrix} 1 & -1 \\ b & b \end{bmatrix}$$

##### Cálculo do Condicionamento de $M$:
Calculamos o produto $M^*M$:
$$M^*M = \begin{bmatrix} 1 & b \\ -1 & b \end{bmatrix} \begin{bmatrix} 1 & -1 \\ b & b \end{bmatrix} = \begin{bmatrix} 1+b^2 & -1+b^2 \\ -1+b^2 & 1+b^2 \end{bmatrix}$$

Buscamos os autovalores $\lambda$ de $M^*M$ através de seu determinante:
$$\det \begin{bmatrix} 1+b^2-\lambda & -1+b^2 \\ -1+b^2 & 1+b^2-\lambda \end{bmatrix} = 0$$

$$(1+b^2-\lambda)^2 - (-1+b^2)^2 = 0$$

Expandindo e simplificando os termos algébricos:
$$(1+b^2)^2 - 2\lambda(1+b^2) + \lambda^2 - (b^4 - 2b^2 + 1) = 0$$

$$b^4 + 2b^2 + 1 - 2\lambda(1+b^2) + \lambda^2 - b^4 + 2b^2 - 1 = 0$$

$$4b^2 - 2\lambda(1+b^2) + \lambda^2 = 0$$

Resolvendo a equação quadrática para $\lambda$ via discriminante ($\Delta$):
$$\Delta = [-2(1+b^2)]^2 - 4(1)(4b^2) = 4(b^4 + 2b^2 + 1) - 16b^2 = 4b^4 - 8b^2 + 4$$

$$\lambda = \frac{2+2b^2 \pm \sqrt{4b^4 - 8b^2 + 4}}{2} = \frac{2+2b^2 \pm 2\sqrt{b^4 - 2b^2 + 1}}{2}$$

$$\lambda = 1 + b^2 \pm (b^2 - 1)$$

O que nos dá os dois autovalores extremantes:
$$\lambda_{\max} = 2b^2 \quad \text{e} \quad \lambda_{\min} = 2$$

Dessa forma, o número de condicionamento da matriz de autovetores (ao quadrado) é dado pela razão destes autovalores:
$$\kappa(M)^2 = \frac{\lambda_{\max}}{\lambda_{\min}} = \frac{2}{2b^2} = \frac{1}{b^2}$$

Como determinado anteriormente, $\frac{1}{b^2} = \kappa(A(\omega))$, o que nos leva à relação final:
$$\kappa(M)^2 = \kappa(A(\omega)) \implies \kappa(M) = \frac{1}{b} = \frac{1}{\omega}$$

```python
ws = np.linspace(0,1, 200)
lista_cond = []

for w in ws:
    Aw = np.array([[0, 1], [-w**2, 0]])
    valores_singulares = np.linalg.svd(Aw, compute_uv=False)
    if valores_singulares.min()==0:
        cond = np.inf
    else:
        cond = valores_singulares.max()/valores_singulares.min()
    lista_cond.append(cond)
```

```python
plt.plot(ws, lista_cond)
plt.title("Condicionamento por $\omega$")
plt.xlabel("$\omega$")
plt.ylabel("$K(A_\omega)$")
plt.yscale("log")
plt.grid()
plt.legend()
plt.show()
```

### Análise do Gráfico: Condicionamento por $\omega$

**O que observamos no gráfico:**
* Em $\omega = 1$, a curva toca exatamente o $10^0$ (ou seja, $\kappa = 1$). 
* Conforme $\omega$ se aproxima de $0$, a curva "explode" exponencialmente para cima, ultrapassando a marca de $10^4$ ($10.000$) e tendendo ao infinito.

**Conclusão Numérica:**
O gráfico é a representação visual exata da função $f(\omega) = 1/\omega^2$. 

1. **A matriz ideal ($\omega = 1$):** Quando $\kappa = 1$, dizemos que a matriz é perfeitamente bem-condicionada (é o caso de matrizes ortogonais). O computador consegue resolver sistemas com essa matriz sem amplificar nenhum erro de arredondamento.
2. **A explosão numérica ($\omega \to 0$):** O crescimento assintótico da curva mostra o sistema se tornando mal condicionado. Perto de $\omega = 0$, qualquer erro microscópico nos dados de entrada (na casa de $10^{-16}$) será multiplicado por um fator gigantesco, destruindo completamente a precisão da resposta do computador.


## Questão 3c)

**Enunciado:** Exponenciais (Bônus): Mostre que $ A(\omega)^2 = -\omega^2 I $. Deduza da série de Taylor para a exponencial que

$e^{tA(\omega)} = \cos(\omega t)\, I + \frac{\sin(\omega t)}{\omega} A(\omega)$.

Calcule $e^{tA(\omega)}$ diretamente no computador e faça o gráfico da norma do erro relativo.

Verifique que, quando $\omega \to 0^+$, o limite da fórmula acima de fato dá $e^{tA(0)}$.

### Demonstração da fórmula para $e^{tA(\omega)}$

Considere

$A(\omega)=\begin{pmatrix}0 & 1\\-\omega^2 & 0\end{pmatrix}$.

Como

$A(\omega)^2=-\omega^2I$,

temos

$A(\omega)^3=A(\omega)A(\omega)^2=-\omega^2A(\omega)$,

$A(\omega)^4=(A(\omega)^2)^2=\omega^4I$.

Por indução, segue que

$A(\omega)^{2k}=(-1)^k\omega^{2k}I$,

e

$A(\omega)^{2k+1}=(-1)^k\omega^{2k}A(\omega)$.

A série de Taylor da exponencial matricial é

$e^{tA(\omega)}=\sum_{k=0}^{\infty}\frac{(tA(\omega))^k}{k!}$.

Separando os termos pares dos ímpares,

$e^{tA(\omega)}=\sum_{k=0}^{\infty}\frac{t^{2k}A(\omega)^{2k}}{(2k)!}+\sum_{k=0}^{\infty}\frac{t^{2k+1}A(\omega)^{2k+1}}{(2k+1)!}$.

Substituindo as expressões para as potências de $A_\omega$,

$e^{tA(\omega)}=\sum_{k=0}^{\infty}\frac{(-1)^k(\omega t)^{2k}}{(2k)!}I+\sum_{k=0}^{\infty}\frac{(-1)^k\omega^{2k}t^{2k+1}}{(2k+1)!}A(\omega)$.

Colocando $A(\omega)/\omega$ em evidência na segunda soma,

$e^{tA(\omega)}=\left(\sum_{k=0}^{\infty}\frac{(-1)^k(\omega t)^{2k}}{(2k)!}\right)I+\frac{A(\omega)}{\omega}\left(\sum_{k=0}^{\infty}\frac{(-1)^k(\omega t)^{2k+1}}{(2k+1)!}\right)$.

Reconhecendo as séries de Taylor de $\cos(x)$ e $\sin(x)$,

$\cos(x)=\sum_{k=0}^{\infty}\frac{(-1)^k x^{2k}}{(2k)!}$,

$\sin(x)=\sum_{k=0}^{\infty}\frac{(-1)^k x^{2k+1}}{(2k+1)!}$,

e tomando $x=\omega t$, obtemos

$e^{tA(\omega)}=\cos(\omega t)\,I+\frac{\sin(\omega t)}{\omega}A(\omega)$.

Portanto,

$\boxed{e^{tA(\omega)}=\cos(\omega t)\,I+\frac{\sin(\omega t)}{\omega}A(\omega)}$.

```python
from scipy.linalg import expm

# Configurações de parâmetros
t = 1.0
ws = np.logspace(-10, 0, 500)  # Escala logarítmica para ver o comportamento bem perto de 0

erros_relativos = []
valores_fórmula_12 = [] # Vamos monitorar o componente (0,1) da matriz para o segundo gráfico
valores_expm_12 = []

for w in ws:
    # 1. Matriz A(w)
    Aw = np.array([[0, 1], [-w**2, 0]])
    
    if w < 1e-12:
    # Evita divisão por zero ou números subnormais usando o limite analítico direto
        exp_analitica = np.array([[np.cos(w*t), t], [0, np.cos(w*t)]]) 
    else:
        term_cos = np.cos(w * t) * np.eye(2)
        term_sin = (np.sin(w * t) / w) * Aw
        exp_analitica = term_cos + term_sin


    # 2. Exponencial direta pelo computador
    exp_computador = expm(t * Aw)
    
    # Cálculo do erro relativo (Norma de Frobenius)
    erro = np.linalg.norm(exp_computador - exp_analitica) / np.linalg.norm(exp_computador)
    erros_relativos.append(erro)
    
    valores_fórmula_12.append(exp_analitica[0, 1])
    valores_expm_12.append(exp_computador[0, 1])

# --- GRÁFICO 1: Norma do Erro Relativo ---
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.loglog(ws, erros_relativos, color='crimson', linewidth=2, label='Erro Relativo')
plt.title("Norma do Erro Relativo ($e^{tA}$)\nComputador vs Fórmula Analítica", fontsize=11, fontweight='bold')
plt.xlabel("Parâmetro $\omega$ (Escala Logarítmica)")
plt.ylabel("Erro Relativo (Norma de Frobenius)")
plt.grid(True, which="both", linestyle=':', alpha=0.5)
plt.axvline(1e-8, color='black', linestyle='--', alpha=0.7, label='Limite de Precisão do Computador')
plt.legend()

# GRÁFICO 2: Verificação do Limite quando w -> 0
# Caso limite teórico e^{tA(0)} = [[1, t], [0, 1]] -> componente [0,1] deve ser igual a t
limite_teorico = t 

plt.subplot(1, 2, 2)
plt.semilogx(ws, valores_fórmula_12, color='blue', linewidth=2.5, label='Fórmula Analítica')
plt.axhline(limite_teorico, color='green', linestyle='--', linewidth=2, label=f'Limite Teórico $e^{{tA(0)}}_{{0,1}} = {t}$')
plt.title("Convergência do Componente $[0,1]$ da Matriz\nquando $\omega \to 0^+$", fontsize=11, fontweight='bold')
plt.xlabel("Parâmetro $\omega$ (Escala Logarítmica)")
plt.ylabel("Valor do Componente $e^{tA}_{0,1}$")
plt.grid(True, which="both", linestyle=':', alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

### Análise do Erro Relativo

O gráfico da esquerda apresenta a norma do erro relativo entre a exponencial calculada numericamente pelo computador e a expressão analítica deduzida acima.

O erro foi definido utilizando a norma de Frobenius e é representado em escala logarítmica em função do parâmetro $\omega$.

Observa-se que:

- Para praticamente todos os valores de $\omega$, o erro permanece entre $10^{-16}$ e $10^{-15}$;
- Esses valores coincidem com a precisão de máquina da aritmética `float64`;
- Não há crescimento sistemático do erro à medida que $\omega$ diminui;
- A linha tracejada indica aproximadamente o limite de precisão do computador, mostrando que a discrepância observada é dominada exclusivamente por erros de arredondamento.

Portanto, os resultados confirmam que a expressão analítica reproduz a exponencial matricial com precisão próxima ao limite imposto pela aritmética de dupla precisão.

### Evidência Numérica do Limite

O gráfico da direita mostra o comportamento do elemento $(0,1)$ da matriz $e^{tA(\omega)}$ à medida que $\omega$ varia.

A linha tracejada representa o valor teórico do limite,

$\left(e^{tA(0)}\right)_{0,1}=1$,

para o caso $t=1$.

Observa-se que:

- Para valores pequenos de $\omega$, o elemento permanece extremamente próximo de 1;
- A curva coincide visualmente com o limite teórico em praticamente toda a região próxima da origem;
- Pequenos desvios aparecem apenas para valores maiores de $\omega$, onde a matriz já está mais distante do caso degenerado $\omega=0$.

Esses resultados confirmam numericamente que

$\lim_{\omega\to0^+}e^{tA(\omega)}=e^{tA(0)}$,

em concordância com a análise teórica.

### Conclusão

Os experimentos mostram que a fórmula fechada

$e^{tA(\omega)}=\cos(\omega t)\,I+\frac{\sin(\omega t)}{\omega}A(\omega)$

é numericamente correta, produzindo erros relativos da ordem da precisão de máquina do formato `float64`.

Além disso, os resultados confirmam que a expressão possui o comportamento limite esperado quando $\omega\to0^+$, convergindo para a exponencial da matriz degenerada $A(0)$. Dessa forma, tanto a dedução analítica quanto sua implementação computacional foram validadas com sucesso.
