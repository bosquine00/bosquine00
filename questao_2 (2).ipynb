## Questão 2a

**Enunciado:** Mostre que RX,d é uma inversa à esquerda de LX, ou seja, RX,d ◦LX = idPd.

**Resposta:**

O método $R_X$ pode ser definido intuitivamente como uma matriz que, quando multiplicada por um vetor $b \in \mathbb{R}^m$, retorna os coeficientes do polinômio pertencente a $\mathcal{P}_d$ que melhor se ajusta aos pontos $x$.

### Passo 1: formato de $L_X$
Considerando a base de polinômios $P_d = \{p_1, ..., p_d\}$, temos:

$$
L_x=\begin{pmatrix}
p_0(x_1) & p_1(x_1) & \cdots & p_d(x_1) \\
p_0(x_2) & p_1(x_2) & \cdots & p_d(x_2) \\
\vdots & \vdots & \ddots & \vdots \\
p_0(x_m) & p_1(x_m) & \cdots & p_d(x_m)
\end{pmatrix}
$$


### Passo 2: encontrar $R_X$

Matematicamente, queremos encontrar o polinômio de grau no máximo $d$ que melhor aproxima os dados $b \in \mathbb{R}^m$ no sentido de mínimos quadrados. Para isso, definimos a matriz de avaliação de Lagrange:

$$
L_X \in \mathbb{R}^{m\times(d+1)},
$$

onde cada entrada é dada por:

$$
(L_X)_{ij}=L_j(x_i).
$$

Se $a \in \mathbb{R}^{d+1}$ representa o vetor de coeficientes do polinômio na base de Lagrange, então o vetor:

$$
L_X a
$$

corresponde às avaliações do polinômio nos pontos $x_0,\dots,x_{m-1}$.

O problema consiste em encontrar o vetor $a$ que minimiza o erro quadrático entre os valores aproximados e os dados observados $b$. Assim, resolvemos:

$$
a=\arg\min_{y\in\mathbb{R}^{d+1}} ||L_Xy-b||_2^2.
$$

Esse é um problema clássico de mínimos quadrados. Geometricamente, estamos projetando ortogonalmente o vetor $b$ sobre o espaço coluna de $L_X$.

As equações normais associadas ao problema são:

$$
L_X^TL_Xa=L_X^Tb.
$$

Quando $L_X$ possui posto completo nas colunas, a solução é dada por:

$$
a=(L_X^TL_X)^{-1}L_X^Tb.
$$

Observamos que:

$$
L_X^+=(L_X^TL_X)^{-1}L_X^T
$$

é precisamente a pseudo-inversa de Moore–Penrose de $L_X$. Portanto:

$$
a=L_X^+b.
$$

Logo, o operador de reconstrução é dado por:

$$
R_X=L_X^+.
$$

Além disso, a matriz:

$$
L_XL_X^+
$$

representa a projeção ortogonal sobre a imagem de $L_X$, isto é, sobre o espaço de todos os vetores que podem ser obtidos a partir de polinômios de grau no máximo $d$ avaliados nos pontos de $X$.

### Passo 3: prova que $R_X$ é inversa à esquerda
Temos que: $R_{X,d} \cdot b = (A^* A)^{-1} A^* b$.

Logo, $R_{X,d} \cdot A = (A^* A)^{-1} A^* A = I$

## Questão 2c

**Enunciado:** Como calcular a matriz de RX,d na base canônica de Pd?

**Resposta:**

Como visto anteriormente, o operador de reconstrução pode ser escrito como:

$$
R_{X,d}=(L_X^*L_X)^{-1}L_X^*.
$$

Entretanto, calcular essa expressão diretamente não é numericamente estável. Para entender o motivo, considere a decomposição em valores singulares (SVD) de uma matriz $A$:

$$
A=U\Sigma V^*.
$$

Então:

$$
A^*A
=
V\Sigma^2V^*.
$$

Isso mostra que os valores singulares de $A^*A$ são os quadrados dos valores singulares de $A$. Consequentemente, o número de condicionamento satisfaz:

$$
\kappa(A^*A)=\kappa(A)^2.
$$

Portanto, ao formar explicitamente a matriz normal

$$
L_X^*L_X,
$$

o condicionamento é elevado ao quadrado, amplificando erros de arredondamento e tornando os cálculos numericamente instáveis.

Por esse motivo, não é recomendável calcular $R_{X,d}$ diretamente através das equações normais. Em vez disso, utilizamos a fatoração QR de $L_X$:

$$
L_X=QR,
$$

Substituindo essa fatoração na expressão de $R_{X,d}$, obtemos:

\begin{align*}
R_{X,d}
&=(L_X^*L_X)^{-1}L_X^* \\
&=((QR)^*(QR))^{-1}(QR)^* \\
&=(R^*Q^*QR)^{-1}R^*Q^* \\
&=(R^*IR)^{-1}R^*Q^* \\
&=(R^*R)^{-1}R^*Q^*.
\end{align*}

Agora utilizamos o fato de que $R$ é inversível:

\begin{align*}
(R^*R)^{-1}
&=R^{-1}(R^*)^{-1}.
\end{align*}

Logo:

\begin{align*}
R_{X,d}
&=R^{-1}(R^*)^{-1}R^*Q^* \\
&=R^{-1}Q^*.
\end{align*}

Assim, obtemos a forma numericamente estável:

$$
R_{X,d}=R^{-1}Q^*.
$$

Essa abordagem evita a formação explícita de $L_X^*L_X$, preservando melhor a estabilidade numérica e reduzindo a propagação de erros computacionais.

### **Passo 2:** analisar variação de d e m

Para isso, foram realizados vários testes: sendo mais preciso $ 2 \le d \le 24$ e $20 \le m \le 280$.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from funcoes_auxiliares import cond, matriz_lagrange
```

```python
dict_lx = {}
dict_rx = {}

for i in range(11):
    dict_lx[30+25*i] = []
    dict_rx[30+25*i] = []

for m in range(30, 281, 25):
    for d in range(2,25):
        x = np.linspace(0,1, m)
        v = np.zeros(shape=(m,d+1))

        v[:,0] = np.ones(m)

        for expoente in range(1,d+1):
            v[:, expoente] = x*v[:, expoente-1]
        condi_v = cond(v)
        rx = np.linalg.pinv(v)
        condi_rx = cond(rx)

        dict_lx[m].append(condi_v.round(2))
        dict_rx[m].append(condi_rx.round(2))

df_lx = pd.DataFrame(dict_lx)
df_lx.index = np.arange(2, 25)
df_lx.columns.name = "m"
df_lx.index.name = "d"
display(df_lx)
```

```python
#PLOTAR OS RESULTADOS ENCONTRADOS

dados_m_1 = df_lx.loc[3]
dados_m_2 = df_lx.loc[4]
dados_m_3 = df_lx.loc[5]
dados_p_1 = df_lx[105]
dados_p_2 = df_lx[130]
dados_p_3 = df_lx[155]

plt.figure()
plt.plot(dados_m_1, label="d=3")
plt.plot(dados_m_2, label="d=4")
plt.plot(dados_m_3, label= "d=5")
plt.title("K(A) x Tamanho m de X")
plt.xlabel("m")
plt.ylabel("K(A)")
plt.legend()
plt.plot()

plt.figure()
plt.plot(dados_p_1, label = "m=105")
plt.plot(dados_p_2, "--", label = "m=130")
plt.plot(dados_p_3, "--",label="m=155")
plt.title("K(A) x Dimensão d do polinômio")
plt.xlabel("d")
plt.ylabel("K(A) - escala log")
plt.yscale("log")
plt.legend()
plt.plot()
```

## Questão 2e

**Enunciado:** Calcule as matrizes de $L_X$ e $R_{X,d}$ usando a base dos polinômios interpoladores de La
grange. Qual o condicionamento das matrizes $L_X$ e $R_{X,d}$ nestas bases? 

**Resposta:**

Utilizando a base dos polinômios interpoladores de Lagrange, a estrutura das matrizes depende diretamente da relação entre o número de pontos avaliados ($m$) e a dimensão do espaço ($d+1$).

### Caso 1: $m = d+1$
Como não há pontos extras, a matriz de avaliação $L_X$ resulta na matriz identidade $I_{d+1}$. Logo, o cálculo da pseudo-inversa $R_{X,d}$ torna-se simplificado.

\begin{align*}
\boldsymbol{L_X} &= \boldsymbol{I_{d+1}} \\
R_{X,d} &= (L_X^T L_X)^{-1} L_X^T \\
&= (I^T I)^{-1} I^T \\
&= (I \cdot I)^{-1} I \\
&= I^{-1} I \\
\boldsymbol{R_{X,d}} &= \boldsymbol{I_{d+1}}
\end{align*}

Na análise do condicionamento em $m = d+1$, temos ambos resultados como uma matriz Identidade, logo o condicionamento é k = 1.

### Caso 2: $m > d+1$
Neste caso, a matriz $L_X$ fica dividida em blocos. O bloco superior é a matriz identidade $I_{d+1}$ (correspondente aos $d+1$ pontos que definem a base de Lagrange) e o bloco inferior $B$ "Blob" $\in \mathbb{R}^{(m-(d+1)) \times (d+1)}$



\begin{align*}
\boldsymbol{L_X} &= \boldsymbol{\begin{pmatrix} I_{d+1} \\ B \end{pmatrix}} \\
R_{X,d} &= (L_X^T L_X)^{-1} L_X^T \\
&= \left( \begin{pmatrix} I & B^T \end{pmatrix} \begin{pmatrix} I \\ B \end{pmatrix} \right)^{-1} \begin{pmatrix} I & B^T \end{pmatrix} \\
&= (I \cdot I + B^T B)^{-1} \begin{pmatrix} I & B^T \end{pmatrix} \\
\boldsymbol{R_{X,d}} &= \boldsymbol{(I_{d+1} + B^T B)^{-1} \begin{pmatrix} I_{d+1} & B^T \end{pmatrix}}
\end{align*}

No condicionamento de $m > d+1$, a matriz $L_X$ possui uma matriz Identidade na parte superior que deixa as colunas praticamente ortogonais, o que nos retorna um condicionamento próximo de 1. Já em $R_{X,d}$, o condicionamento sofre influência da invertibilidade da matriz $(I_{d+1} + B^T B)$. Podemos analisar essa estabilidade observando o espectro de autovalores do sistema. Pelo teorema espectral, os autovalores dessa matriz será $\lambda_i \ge 0$.
A operação de somar a matriz identidade $I_{d+1}$ aplica um deslocamento à toda matriz. Consequentemente, os autovalores do termo a ser invertido tornam-se limitados: $\lambda_i(I_{d+1} + B^T B) \ge 1$. Logo, o número de condicionamento está ligado à razão entre o maior e o menor autovalor, onde o cálculo da inversa não amplifica os ruídos numéricos.


Portanto, o condicionamento $\kappa(R_{X,d})$ nesta base permanece limitado e estritamente pequeno, denotando um sistema ótimo e altamente estável.

```python
#TENHO QUE GERAR UMA MATRIZ DE LAGRANGE
#PARA M+1>D, TEMOS LX = [I, BLOB]
#I TEM DIMENSÃO D+1, O RESTO SE CALCULA A PARTIR DO POLINOMIO


dict_lx = {}

for i in range(11):
    dict_lx[30 + 25*i] = []

for m in range(30, 281, 25):

    for d in range(2, 25):

        try:

            # mesmos pontos usados em calcular_lagrange
            pontos_base = np.linspace(-1, 1, d + 1)

            extra = np.linspace(
                -1,
                1,
                m - (d + 1)
            )

            pontos_destino = np.concatenate(
                [pontos_base, extra]
            )

            lx = matriz_lagrange(
                pontos_destino,
                pontos_base
            )

            s = np.linalg.svd(
                lx,
                compute_uv=False
            )

            condi_lx = s[0] / s[-1]

            dict_lx[m].append(condi_lx)

        except Exception:

            dict_lx[m].append(np.nan)


```

```python
df = pd.DataFrame(dict_lx)
df.index = np.arange(2, 25)
df.columns.name = "m"
df.index.name = "d"
display(df)
```

```python
#PLOTAR OS RESULTADOS ENCONTRADOS

dados_m_1 = df.loc[3]
dados_m_2 = df.loc[4]
dados_m_3 = df.loc[5]
dados_p_1 = df[105]
dados_p_2 = df[130]
dados_p_3 = df[155]

plt.figure()
plt.plot(dados_m_1, label="d=3")
plt.plot(dados_m_2, label="d=4")
plt.plot(dados_m_3, label= "d=5")
plt.title("K(A) x Tamanho m de X")
plt.xlabel("m")
plt.ylabel("K(A)")
plt.grid(True)
plt.legend()
plt.plot()

plt.figure()
plt.plot(dados_p_1, label = "m=30")
plt.plot(dados_p_2, label = "m=130")
plt.plot(dados_p_3,label="m=155")
plt.title("K(A) x Tamamho p do polinômio")
plt.xlabel("d")
plt.ylabel("K(A)")
plt.yscale("log")
plt.grid(True)
plt.legend()
plt.plot()
```

## Questão 2f

**Enunciado:** Compare a complexidade computacional de calcular $L_X$ e $R_{X,d}$ usando as bases canônica
e de Lagrange com o condicionamento destas matrizes.

**Resposta:**

**Complexidade e Condicionamento na Base Canônica:** 

**Cálculo de $L_X$:** Na matriz de Vandermonde, temos que calcular as potências $x^0, x^1, \dots, x^d$ para $m$ pontos. Logo, a complexidade é $\mathcal{O}(md)$

**Cálculo de $R_{X,d}$:** Utilizando a pseudo-inversa via equações normais $(L_X^T L_X)^{-1} L_X^T$, o custo é dominado pela multiplicação $L_X^T L_X$, que exige $\mathcal{O}(m d^2)$ operações, seguida da inversão de uma matriz $(d+1) \times (d+1)$ que custa $\mathcal{O}(d^3)$. Como $m > d$, a complexidade geral $\mathcal{O}(md^2)$.
 

Apesar de ser um algoritmo direto o condicionamento $\kappa(L_X)$ cresce exponencialmente com $d$. Portanto, a base canônica é computacionalmente barata, mas numericamente inútil para graus elevados.


**Complexidade e Condicionamento na Base de Lagrange:**

**Cálculo de $L_X$:** Para $m = d+1$, $L_X = I_{d+1}$, com custo $\mathcal{O}(d)$. Para $m > d+1$, avaliamos os polinômios nos pontos restantes utilizando pesos pré-computados (fórmula baricêntrica). O pré-cálculo dos pesos exige $\mathcal{O}(d^2)$, e a avaliação de cada uma das $m - (d+1)$ linhas restantes custa $\mathcal{O}(d)$. A complexidade final é $\mathcal{O}(d^2 + md)$.

**Cálculo de $R_{X,d}$:** Como $L_X$ possui a estrutura de blocos com $I_{d+1}$ no topo e $B$ na base, na matriz $(L_X^T L_X) = I + B^T B$, multiplicação $B^T B$ exige $\mathcal{O}((m-d)d^2) \approx \mathcal{O}(md^2)$. Somada à inversão $\mathcal{O}(d^3)$, a complexidade mantém-se em $\mathcal{O}(md^2)$.

O condicionamento é extremamente baixo e estável, sendo sustentado pelo bloco identidade $I_{d+1}$, permitindo estabilidade numérica mesmo para altos valores de $d$.


Logo, é possível constatar que as operações matemáticas para construir e inverter os sistemas são da mesma ordem em ambas as abordagens: $\mathcal{O}(md)$ para avaliação e $\mathcal{O}(md^2)$ para regressão.No entanto, enquanto a base canônica leva muito tempo para construir uma matriz quase singular, a base de Lagrange utiliza o mesmo processamento para construir um sistema com autovalores limitados por $1$. Portanto, a base de Lagrange entrega estabilidade numérica elevada, possuindo a mesma complexidade algorítmica da base canônica.
