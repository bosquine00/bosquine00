## Questão 3a)

Seja $L_Y \circ R_{X,d} : \mathbb{R}^m \to \mathbb{R}^n$, os polinômios pertencem a $P_d$ que possui dimensão $d + 1$, como $m \ge d + 1$, então a imagem de $R_{X,d}$ cobre todo o espaço de $P_d$ podendo gerar qualquer polinômio desse conjunto, o que não nos retira graus de liberdade. Como a primeira etapa nos entrega o espaço $\mathcal{P}_d$ completo, o posto da aplicação composta será exatamente o posto da matriz $L_Y$ agindo sobre um espaço de dimensão $d+1$. A matriz $L_Y$ avalia os polinômios em $n$ pontos distintos. Aqui, o posto vai depender de quem é maior, $n$ ou $d+1$:

**Se $n \ge d+1$:** Nesse caso, estamos avaliando o polinômio com mais pontos do que coeficientes. Como os pontos de Y são únicos, então as colunas da matriz serão **Linearmente Independentes**, logo o posto será d + 1.

**Se $n < d+1$:** O espaço de destino $\mathbb{R}^n$ é menor do que a dimensão do polinômio, não tem como a imagem ter um número de dimensões maior do que o espaço em que ela vive. Como a avaliação em pontos distintos cobre todas as combinações possíveis em $\mathbb{R}^n$ (sobrejetividade), o posto será limitado a n.

Portanto, como a dimensão final é sempre retida pela menor capacidade encontrada ao longo desse fluxo algébrico, concluímos que: $$\text{posto}(L_Y \circ R_{X,d}) = \min(n, d+1)$$

## Questão 3b)

A aplicação composta $L_Y \circ R_{X,d}$ é estritamente independente da base escolhida para representar o espaço polinomial $\mathcal{P}_d$ devido à unicidade geométrica da projeção ortogonal. Podemos demonstrar isto analisando a natureza de cada operador:

**A Unicidade do Polinômio de Regressão ($R_{X,d}$):** O operador $R_{X,d}$ resolve um problema de mínimos quadrados, buscando o polinômio que minimiza a norma euclidiana $\|p(X) - b\|_2^2$. Pelo Teorema da Projeção Ortogonal em espaços de dimensão finita, a projeção de um vetor sobre um subespaço (neste caso, o subespaço gerado por $\mathcal{P}_d$ nos pontos $X$) é única. Assim, a curva polinomial $p(x)$ gerada é invariável, sendo rigorosamente a mesma curva geométrica não importando a base utilizada.

**A Mudança de Coordenadas:** Escolher bases diferentes (como Canônica, Lagrange ou Chebyshev) altera apenas o sistema de coordenadas computacional, ou seja, altera as matrizes intermediárias e o vetor de coeficientes calculado. No entanto, a combinação linear desses coeficientes com as respectivas funções de base resultará no mesmo polinômio $p(x)$.

**A Avaliação Determinística ($L_Y$):** A segunda etapa da composição, o operador $L_Y$, consiste em avaliar este polinômio $p(x)$ nos pontos do conjunto $Y$. Como a função $p(x)$ é única e perfeitamente definida, os valores de saída $p(y_i)$ dependerão exclusivamente da posição da curva nesses pontos, tornando o vetor resultante em $\mathbb{R}^n$ imune à escolha da representação algébrica intermediária.


Portanto, a composição é uma propriedade intrínseca do espaço vetorial $\mathcal{P}_d$ em si, e não das bases particulares utilizadas para o parametrizar.

## Questão 3c)

A composição

$$
T = L_Y \circ R_{X,d} : \mathbb{R}^m \to \mathbb{R}^n
$$

pode ser interpretada como um operador de aproximação e reamostragem. Dado um vetor de dados

$$
b=(b_1,\dots,b_m)\in\mathbb{R}^m,
$$

o operador $R_{X,d}$ procura o polinômio de grau no máximo $d$ que melhor ajusta esses dados no sentido de mínimos quadrados.

Para entender essa construção, considere o espaço

$$
\operatorname{Im}(L_X),
$$

formado por todos os vetores que podem ser obtidos pela avaliação de polinômios de grau no máximo $d$ nos pontos de $X$. Em geral, um vetor arbitrário $b\in\mathbb{R}^m$ não pertence a esse espaço. Nesse caso, $R_{X,d}$ determina o polinômio $p\in\mathcal P_d$ cujas avaliações nos pontos de $X$ produzem o vetor de $\operatorname{Im}(L_X)$ mais próximo de $b$.

Em outras palavras, procuramos um polinômio $p$ tal que

$$
p(x_i)\approx b_i,
\qquad i=1,\dots,m,
$$

e cuja aproximação minimize o erro quadrático total.

Uma vez determinado esse polinômio, o operador $L_Y$ o avalia em um novo conjunto de pontos

$$
Y=\{y_1,\dots,y_n\}.
$$

Assim, o operador composto $T=L_Y\circ R_{X,d}$ realiza duas etapas:

1. projeta os dados $b$ sobre o espaço dos vetores gerados por polinômios de grau no máximo $d$;
2. avalia o polinômio obtido nos pontos de $Y$.

Portanto, $T$ pode ser visto como um mecanismo que recebe dados observados nos pontos de $X$, constrói a melhor aproximação polinomial (interpolação) desses dados e, em seguida, transfere essa aproximação para os pontos de $Y$.

Observe que, se tomarmos $Y=X$ e $R_{X,d}=L_X^{+}$, então

$$
T=L_XL_X^{+}.
$$

Sabe-se que, $L_XL_X^{+}$ é a projeção ortogonal sobre a imagem de $L_X$.

Em particular, se $b \in \operatorname{Im}(L_X)$, então

$$
Tb=L_XL_X^{+}b=b.
$$

Ou seja, quando $b$ corresponde às avaliações de algum polinômio $p \in \mathcal{P}_d$ nos pontos de $X$, o operador recupera exatamente esse vetor de avaliações.

Por outro lado, se $b$ for arbitrário e não pertencer à imagem de $L_X$, então,

$$
Tb \neq b.
$$

Nesse caso, $Tb$ é a projeção ortogonal de $b$ sobre $\operatorname{Im}(L_X)$, correspondendo ao polinômio de mínimos quadrados associado aos dados.

**Resposta:** 

Antes de analisar os resultados numéricos, é importante observar uma propriedade teórica da matriz composta $T = L_Y \circ R_{X,d}$.

A matriz que representa T depende da escolha de uma base para o espaço $\mathcal P_d$. Considere a base canônica

{${1,x,x^2,\ldots ,x^d}$}

e uma base de Lagrange

$\{\ell_0,\ell_1,\ldots,\ell_d\}$

Como ambas são bases do mesmo espaço vetorial $\mathcal P_d$, existe uma matriz inversível de mudança de base $M$ tal que

$L_X^{\text{lag}} = L_X^{\text{can}} M
\qquad\text{e}\qquad
L_Y^{\text{lag}} = L_Y^{\text{can}} M$

Consequentemente,

$T_{\text{lag}}=L_Y^{\text{lag}} \left(L_X^{\text{lag}}\right)^+=
L_Y^{\text{can}} 
\left(L_X^{\text{can}}\right)^+=T_{\text{can}}$

Portanto, matematicamente, a mudança de base não altera $T$. Ou seja, os valores singulares de T deveriam ser exatamente os mesmos, independentemente da base utilizada. Entretanto, ao computar essas matrizes numericamente, surgem erros de arredondamento decorrentes da precisão finita da máquina. Como as bases possuem condicionamentos distintos, esses erros podem ser amplificados de maneiras diferentes. Assim, embora as matrizes calculadas nas bases canônica e de Lagrange representem o mesmo operador matemático, elas podem apresentar pequenas diferenças numéricas.

Dessa forma, o comportamento esperado nos experimentos é que os valores singulares obtidos pelas duas abordagens sejam idênticos ou muito próximos, com discrepâncias atribuídas exclusivamente aos erros numéricos introduzidos durante o processo de cálculo.
Para responder corretamente a questão serão realizados um conjunto de testes (para diferentes valores de m,n,d) e, depois, será plotado um gráfico que mostra os valores singulares em que a composta de cada base chegou.

Antes de fazermos o plot, será dada uma breve explicação sobre a estrutura de $T$. Temos que $T_{(n,m)}$, uma vez que $L_Y{(n,d)}$ e $R_X{(d,m)}$. Logo, uma matriz $T$ possui $ \min (n,m)$ valores singulares.

**OBSERVAÇÃO**: para a resolução dessa questão, será analisada apenas os d primeiros valores singulares da composta, isto é, os valores singulares que matematicamente podem ser maiores que 0. Por esses de arredondamento, no cálculo dos valores singures $ \sigma_i$, para $i>d$, é comum termos $ \sigma_i >0$. 

```python
import numpy as np
import pandas as pd
from funcoes_auxiliares import *
import matplotlib.pyplot as plt
```

```python
# CASOS DE TESTE
valores_m = [10, 25,50,75,100, 500, 750, 1000]
valores_n = [10, 25,50,75,100, 500, 750, 1000]
valores_d = [4, 6,8,10,12,14,16,18, 20, 50, 100]

resultados = []

for m in valores_m:
    for n in valores_n:
        d_max = min(m, n) - 1
        for d in valores_d:
            if d <= d_max:
                try:
                    X = np.linspace(-1, 1, m)
                    Y = np.linspace(-1, 1, n)
                    idx = np.linspace(0, m - 1, d + 1, dtype=int)
                    X_base = X[idx]


                    # BASE CANÔNICA
                    LX_can = matriz_vander(X, d)
                    LY_can = matriz_vander(Y, d)
                    cond_LX_can = np.linalg.cond(LX_can)
                    RX_can = np.linalg.pinv(LX_can)
                    T_can = LY_can @ RX_can
                    S_can = np.linalg.svd(T_can, compute_uv=False)
                    tol_can = (
                        max(T_can.shape)
                        * np.max(S_can)
                        * np.finfo(float).eps
                    )

                    sigma_max_can = np.max(S_can)
                    sigma_min_nz_can = np.min(
                        S_can[S_can > tol_can]
                    )

                    cond_T_can = (
                        sigma_max_can
                        / sigma_min_nz_can
                    )

                    # BASE DE LAGRANGE
                    LX_lag = matriz_lagrange(
                        X,
                        X_base
                    )
                    LY_lag = matriz_lagrange(
                        Y,
                        X_base
                    )
                    cond_LX_lag = np.linalg.cond(LX_lag)
                    RX_lag = np.linalg.pinv(LX_lag)
                    T_lag = LY_lag @ RX_lag
                    S_lag = np.linalg.svd(
                        T_lag,
                        compute_uv=False
                    )
                    tol_lag = (
                        max(T_lag.shape)
                        * np.max(S_lag)
                        * np.finfo(float).eps
                    )
                    sigma_max_lag = np.max(S_lag)
                    sigma_min_nz_lag = np.min(
                        S_lag[S_lag > tol_lag]
                    )
                    cond_T_lag = (
                        sigma_max_lag
                        / sigma_min_nz_lag
                    )


                    # COMPARAÇÃO
                    erro_maximo = np.max(
                        np.abs(T_can - T_lag)
                    )
                    erro_fro = np.linalg.norm(
                        T_can - T_lag,
                        ord="fro"
                    )
                    resultado = {
                        "m": m,
                        "n": n,
                        "d": d,
                        "cond_LX_can": cond_LX_can,
                        "cond_LX_lag": cond_LX_lag,
                        "cond_T_can": cond_T_can,
                        "cond_T_lag": cond_T_lag,
                        "sigma_max_can": sigma_max_can,
                        "sigma_min_can": sigma_min_nz_can,

                        "sigma_max_lag": sigma_max_lag,
                        "sigma_min_lag": sigma_min_nz_lag,
                        "erro_maximo": erro_maximo,
                        "erro_fro": erro_fro,
                    }

                    resultados.append(resultado)

                except Exception as e:
                    resultados.append(
                        {
                            "m": m,
                            "n": n,
                            "d": d,
                            "cond_canonica": np.nan,
                            "cond_lagrange": np.nan,
                            "erro_maximo": str(e),
                        }
                    )


# DATAFRAME FINAL
df = pd.DataFrame(resultados)

pd.set_option("display.max_rows", None)
pd.set_option("display.float_format", "{:.3e}".format)

df
```

Entretanto, o comportamento muda drasticamente nos três últimos experimentos, onde $d=100$. Enquanto os valores singulares associados à base canônica permanecem próximos de 1 durante boa parte do espectro e decaem apenas nos índices finais, os valores singulares obtidos pela base de Lagrange apresentam uma queda progressiva ao longo de todo o espectro, atingindo valores extremamente pequenos, da ordem de $10^{-18}$.

À primeira vista, esse resultado poderia sugerir que a base canônica se torna superior para graus elevados, o que contrairia a questão 2f). No entanto, essa interpretação é incorreta. O que ocorre é um fenômeno de instabilidade numérica associado à interpolação polinomial em nós igualmente espaçados, conhecido como **Fenômeno de Runge**. O Fenômeno de Runge descreve o comportamento observado quando se tenta interpolar funções utilizando polinômios de grau elevado em pontos uniformemente distribuídos. À medida que o grau do polinômio aumenta, surgem oscilações cada vez maiores próximas às extremidades do intervalo de interpolação. Geometricamente, isso acontece porque os polinômios interpoladores precisam passar exatamente pelos valores prescritos nos nós. Em graus elevados, pequenas alterações nos dados podem produzir grandes variações no polinômio resultante, principalmente nas regiões próximas de $x=-1$ e $x=1$.

A base de Lagrange é construída explicitamente a partir dos polinômios interpoladores associados aos nós escolhidos. Portanto, ela está diretamente ligada ao problema de interpolação.
Quando os nós são uniformes e d é grande, os polinômios de Lagrange passam a apresentar amplitudes muito elevadas nas extremidades do intervalo. Como consequência, a matriz de avaliação
$L_X$
torna-se extremamente mal condicionada. Pequenos erros de arredondamento são amplificados durante o cálculo da pseudoinversa e contaminam os menores valores singulares da composta.
Em outras palavras, a base de Lagrange sofre diretamente as consequências do Fenômeno de Runge porque seus vetores de base são exatamente os polinômios responsáveis pela interpolação.

A base canônica também não está imune a problemas numéricos. De fato, matrizes de Vandermonde são conhecidas por serem mal condicionadas para graus elevados. Entretanto, os erros manifestam-se de maneira diferente. A representação monomial não incorpora explicitamente os polinômios interpoladores e, por isso, nem sempre evidencia imediatamente as oscilações produzidas pela interpolação. O fato de o espectro singular da base canônica parecer mais estável nesses experimentos não significa que ela seja numericamente melhor; significa apenas que os erros estão distribuídos de forma diferente na representação matricial.

Portanto, a divergência observada nos experimentos com $d=100$ não indica que a base canônica passou a ser superior. Ela revela que o processo de interpolação em nós uniformes entrou em um regime dominado pelos efeitos do Fenômeno de Runge e pelos erros de arredondamento associados ao condicionamento das matrizes envolvidas.

Assim, os gráficos confirmam duas conclusões importantes:

1. Para graus moderados, as bases canônica e de Lagrange produzem praticamente o mesmo espectro singular, como previsto pela teoria.
2. Para graus elevados em nós uniformes, os efeitos do Fenômeno de Runge tornam-se dominantes e a representação em Lagrange passa a sofrer forte degradação numérica, afetando principalmente os menores valores singulares.

### Generalização dos Parâmetros no Infinito ($\to \infty$)

Ao extrapolarmos os parâmetros analiticamente para o infinito, obtemos o seguinte comportamento assintótico:

* **$n \to \infty$ (Não prejudica):** O condicionamento se estabiliza. Aumentar a densidade de pontos de avaliação no alvo apenas refina a discretização da função contínua resultante, convergindo para o operador contínuo.
* **$m \to \infty$ (Não prejudica):** O condicionamento permanece controlado e estável. Aumentar os pontos de amostragem originais melhora a robustez da aproximação por mínimos quadrados (pseudo-inversa), sem causar explosão de erro.

O problema torna-se severamente mal-condicionado devido à quase dependência linear das colunas da matriz de avaliação (especialmente na base canônica) e às oscilações violentas associadas à interpolação polinomial de alto grau em nós uniformes, tornando o processo numericamente instável na prática.

### Item d) para pontos de Chebyshev

A definição algébrica e geométrica da composição $T = L_Y \circ R_{X,d}$ permanece estritamente a mesma: o operador continua atuando como um projetor ortogonal dos dados sobre $\operatorname{Im}(L_X)$, seguido de uma avaliação em $Y$. 

Ao utilizar uma malha equiespaçada, o operador projetava os dados de forma que o erro nas extremidades do intervalo $[-1, 1]$ tendia a crescer descontroladamente para graus elevados, o que nomemos como "Fenômeno de Runge". 

Ao utilizar os pontos de Chebyshev, o operador passa a distribuir o erro de aproximação de maneira quase uniforme ao longo de todo o intervalo. Em termos analíticos, esta escolha de $X$ aproxima a nossa projeção de mínimos quadrados (norma $L_2$) da melhor aproximação polinomial possível no sentido de minimizar o erro máximo (norma $L_\infty$). A Constante de Lebesgue do operador $R_{X,d}$ cai de um crescimento exponencial para um crescimento logarítmico brando $\mathcal{O}(\ln m)$, tornando o mecanismo de interpolação robusto.

```python
valores_m = [10, 25,50,75,100, 500, 750, 1000]
valores_n = [10, 25,50,75,100, 500, 750, 1000]
valores_d = [4, 6,8,10,12,14,16,18, 20, 50, 100]

resultados_chebyshev = []
for m in valores_m:
    for n in valores_n:
        d_max = min(m, n) - 1
        for d in valores_d:
            if d <= d_max:
                try:
                    indices = np.arange(1, m + 1)
                    X = np.cos(
                        (2 * indices - 1)
                        / (2 * m)
                        * np.pi
                    )
                    Y = np.linspace(-1, 1, n)

                    # Base canônica
                    LX_can = matriz_vander(X, d)
                    LY_can = matriz_vander(Y, d)
                    RX_can = np.linalg.pinv(LX_can)
                    T_can = LY_can @ RX_can
                    dados_can = estatisticas_svd(T_can)

                    # Base de Lagrange
                    k = np.arange(d + 1)
                    X_base = np.cos(
                        (2 * k + 1)
                        * np.pi
                        / (2 * (d + 1))
                    )
                    LX_lag = matriz_lagrange(X, X_base)
                    LY_lag = matriz_lagrange(Y, X_base)
                    RX_lag = np.linalg.pinv(LX_lag)
                    T_lag = LY_lag @ RX_lag
                    dados_lag = estatisticas_svd(T_lag)

                    # Comparações
                    S_can = np.linalg.svd(
                        T_can,
                        compute_uv=False
                    )
                    S_lag = np.linalg.svd(
                        T_lag,
                        compute_uv=False
                    )
                    erro_fro = np.linalg.norm(
                        T_can - T_lag,
                        ord="fro"
                    )
                    erro_max = np.max(
                        np.abs(T_can - T_lag)
                    )
                    resultados_chebyshev.append({
                        "m": m,
                        "n": n,
                        "d": d,
                        "cond_LX_can":
                            np.linalg.cond(LX_can),
                        "cond_LX_lag":
                            np.linalg.cond(LX_lag),
                        "sigma_max_can":
                            dados_can["sigma_max"],
                        "sigma_min_can":
                            dados_can["sigma_min_nz"],
                        "cond_T_can":
                            dados_can["cond_T"],
                        "sigma_max_lag":
                            dados_lag["sigma_max"],
                        "sigma_min_lag":
                            dados_lag["sigma_min_nz"],
                        "cond_T_lag":
                            dados_lag["cond_T"],
                        "erro_fro":
                            erro_fro,
                        "erro_max":
                            erro_max,
                    })

                except Exception as e:
                    resultados_chebyshev.append({
                        "m": m,
                        "n": n,
                        "d": d,
                        "cond_LX_can": np.nan,
                        "cond_LX_lag": np.nan,
                        "sigma_max_can": np.nan,
                        "sigma_min_can": np.nan,
                        "cond_T_can": np.nan,
                        "sigma_max_lag": np.nan,
                        "sigma_min_lag": np.nan,
                        "cond_T_lag": np.nan,
                        "erro_fro": np.nan,
                        "erro_max": str(e),
                    })

df_chebyshev = pd.DataFrame(resultados_chebyshev)
pd.set_option("display.float_format", "{:.3e}".format)
df_chebyshev
```

```python
casos_pequenos = [
    (50, 50, 10),
    (25, 75, 20),
    (75,25,20),
]
plt.figure()
fig, axes = plt.subplots(
    1,
    3,
    figsize=(15,4),
    constrained_layout=True
)
for ax, (m, n, d) in zip(axes, casos_pequenos):
    T_can = construir_operador(
        m, n, d,
        base="vander",
        metodo="chebyshev"
    )
    T_lag = construir_operador(
        m, n, d,
        base="lagrange",
        metodo="chebyshev"
    )
    S_can = np.linalg.svd(
        T_can,
        compute_uv=False
    )[:d+1]
    S_lag = np.linalg.svd(
        T_lag,
        compute_uv=False
    )[:d+1]
    ax.semilogy(
        np.arange(1, d+2),
        S_can,
        "o-",
        label="Canônica"
    )
    ax.semilogy(
        np.arange(1, d+2),
        S_lag,
        "s-",
        label="Lagrange"
    )
    ax.legend()
    ax.set_title(f"m=n={m}, d={d}")
    ax.set_xlabel("Índice singular")
    ax.grid(True, which="both")
axes[0].set_ylabel("Valor singular")
fig.suptitle(
    r"Casos pequenos: $\sigma_i$ Canônico x Lagrange",
    fontsize=14,
    fontweight="bold"
)
plt.show()

casos_grandes = [
    (750, 500, 100),
    (500,750, 100),
    (1000,1000,100)
]
plt.figure()
fig, axes = plt.subplots(
    1,
    3,
    figsize=(15,4),
    constrained_layout=True
)
for ax, (m, n, d) in zip(axes, casos_grandes):
    T_can = construir_operador(
        m, n, d,
        base="vander",
        metodo="chebyshev"
    )
    T_lag = construir_operador(
        m, n, d,
        base="lagrange",
        metodo="chebyshev"
    )
    S_can = np.linalg.svd(
        T_can,
        compute_uv=False
    )[:d+1]
    S_lag = np.linalg.svd(
        T_lag,
        compute_uv=False
    )[:d+1]
    ax.semilogy(
        np.arange(1, d+2),
        S_can,
        label="Canônica"
    )
    ax.semilogy(
        np.arange(1, d+2),
        S_lag,
        label="Lagrange"
    )
    ax.set_title(f"m=n={m}, d={d}")
    ax.set_xlabel("Índice singular")
    ax.grid(True, which="both")
    ax.legend()

axes[0].set_ylabel("Valor singular")
fig.suptitle(
    r"Casos grandes: $\sigma_i$ Canônico x Lagrange",
    fontsize=14,
    fontweight="bold"
)
plt.show()
```

Os pontos de Chebyshev adensam a amostragem nas extremidades do intervalo, mitigando o Fenômeno de Runge e mantendo as funções de base bem condicionadas.

### Generalização dos Parâmetros no Infinito ($\to \infty$)

* **$n \to \infty$ (Não prejudica):** O condicionamento estabiliza. Aumentar a resolução do grid de avaliação no alvo apenas refina a representação contínua do operador, sem inflacionar o erro numérico.
* **$m \to \infty$ (Não prejudica):** O condicionamento permanece controlado. Aumentar a quantidade de nós de Chebyshev melhora a robustez da pseudo-inversa por mínimos quadrados, convergindo perfeitamente para o limite contínuo.
* **$d \to \infty$ (Depende estritamente da base utilizada):**
  * Na **Base Canônica**, o condicionamento vai para o infinito ($\kappa \to \infty$) de forma quase imediata, tornando o problema intratável.
  * Na **Base de Lagrange com nós de Chebyshev**, o crescimento de $\kappa$ em função de $d$ é logarithmicamente lento ou polinomial de grau muito baixo. Portanto, apesar de no infinito o condicionamento ainda explodir, **$d$ pode crescer substancialmente** para d grande sem destruir a estabilidade numérica do problema, permitindo aproximações polinomiais de alta ordem que seriam impossíveis com nós uniformes.

## 3g)
**Enunciado:** Extrapolação 1. Considere $X = linspace(−1,1,m)$, e $Y = linspace(−1.5,1.5,n)$.
O que acontece agora?

**Resposta:**

Ao alterar o conjunto de avaliação para $Y = \text{linspace}(-1.5, 1.5, n)$, mantendo o domínio de ancoragem do modelo restrito a $X \subset [-1, 1]$, a aplicação composta $T = L_Y \circ R_{X,d}$ passa a atuar como um operador de extrapolação nas regiões marginais $[-1.5, -1)$ e $(1, 1.5]$, que ocasiona duas consequências numéricas e geométricas.

**Divergência Polinomial Assintótica:** Pela própria natureza analítica, qualquer polinômio de grau $d \ge 1$ tende a $\pm \infty$ à medida que $x$ aumenta. Fora do espaço $[-1, 1]$, o polinômio perde a sua restrição interpoladora. O erro de aproximação deixa de ser limitado e cresce a taxas elevadas, dominado pelo termo $x^d$.

**Degradação do Condicionamento:** A matriz composta $T$ terá a sua norma consideravelmente amplificada. As linhas da matriz de avaliação $L_Y$ associadas aos pontos extrapolados conterão potências (na base canônica) ou produtórios (na base de Lagrange) com valores absolutos $|x| > 1$. Isto gera uma amplificação de qualquer erro numérico ou imprecisão dos coeficientes calculados, corrompendo a estabilidade global do sistema.

```python
valores_m = [10, 25,50,75,100, 500, 750, 1000]
valores_n = [10, 25,50,75,100, 500, 750, 1000]
valores_d = [4, 6,8,10,12,14,16,18, 20, 50, 100]
resultados_extra = []

for m in valores_m:
    for n in valores_n:
        d_max = min(m, n) - 1
        for d in valores_d:
            if d <= d_max:
                try:
                    indices = np.arange(1, m + 1)
                    X = np.linspace(-1,1,m)
                    Y = np.linspace(-1.5, 1.5, n)

                    # Base canônica
                    LX_can = matriz_vander(X, d)
                    LY_can = matriz_vander(Y, d)
                    RX_can = np.linalg.pinv(LX_can)
                    T_can = LY_can @ RX_can
                    dados_can = estatisticas_svd(T_can)

                    # Base de Lagrange
                    idx = np.linspace(0, m - 1, d + 1, dtype=int)
                    X_base = X[idx]
                    LX_lag = matriz_lagrange(X, X_base)
                    LY_lag = matriz_lagrange(Y, X_base)
                    RX_lag = np.linalg.pinv(LX_lag)
                    T_lag = LY_lag @ RX_lag
                    dados_lag = estatisticas_svd(T_lag)

                    # Comparações
                    S_can = np.linalg.svd(
                        T_can,
                        compute_uv=False
                    )
                    S_lag = np.linalg.svd(
                        T_lag,
                        compute_uv=False
                    )
                    erro_fro = np.linalg.norm(
                        T_can - T_lag,
                        ord="fro"
                    )
                    erro_max = np.max(
                        np.abs(T_can - T_lag)
                    )
                    resultados_extra.append({
                        "m": m,
                        "n": n,
                        "d": d,
                        "cond_LX_can":
                            np.linalg.cond(LX_can),
                        "cond_LX_lag":
                            np.linalg.cond(LX_lag),
                        "sigma_max_can":
                            dados_can["sigma_max"],
                        "sigma_min_can":
                            dados_can["sigma_min_nz"],
                        "cond_T_can":
                            dados_can["cond_T"],
                        "sigma_max_lag":
                            dados_lag["sigma_max"],
                        "sigma_min_lag":
                            dados_lag["sigma_min_nz"],
                        "cond_T_lag":
                            dados_lag["cond_T"],
                        "erro_fro":
                            erro_fro,
                        "erro_max":
                            erro_max,
                        "norma_T_can":
                            np.linalg.norm(T_can, 2),
                        "norma_T_lag":
                            np.linalg.norm(T_lag, 2),
                    })

                except Exception as e:

                    resultados_extra.append({
                        "m": m,
                        "n": n,
                        "d": d,
                        "cond_LX_can": np.nan,
                        "cond_LX_lag": np.nan,
                        "sigma_max_can": np.nan,
                        "sigma_min_can": np.nan,
                        "cond_T_can": np.nan,
                        "sigma_max_lag": np.nan,
                        "sigma_min_lag": np.nan,
                        "cond_T_lag": np.nan,
                        "erro_fro": np.nan,
                        "erro_max": str(e),
                        "norma_T_can": np.nan,
                        "norma_T_lag": np.nan,

                    })

df_extra_1 = pd.DataFrame(resultados_extra)
```

## 3h)
**Enunciado:** Adianta usar os pontos de Chebyshev em X?

**Resposta:**

A substituição do conjunto $X$ pelos pontos de Chebyshev não diminui o problema da extrapolação, atuando apenas como um solucionador estrito de interpolação.
A eficácia dos pontos de Chebyshev (raízes do polinômio de Chebyshev $T_d(x)$) vem do teorema minimax, minimizando a Constante de Lebesgue exclusivamente no intervalo fechado $[-1, 1]$. Dentro destas fronteiras, o erro da norma não explode.
No entanto, fora do intervalo $[-1, 1]$, os polinômios de Chebyshev possuem uma taxa de crescimento absoluto mais rápida entre todos os polinômios do mesmo grau que são limitados por $1$ na origem. O valor assintótico de $|T_d(x)|$ para $|x| > 1$ cresce numa ordem de $\mathcal{O}\left((|x| + \sqrt{x^2-1})^d\right)$.
Portanto, ao avaliar a composição no domínio estendido $Y \in [-1.5, 1.5]$, a malha de Chebyshev garantirá um ajuste quase perfeito e estável no intervalo $[-1, 1]$, mas sofrerá uma divergência ainda muito maior logo após cruzar os limites do domínio referencial.


```python
valores_m = [10, 25,50,75,100, 500, 750, 1000]
valores_n = [10, 25,50,75,100, 500, 750, 1000]
valores_d = [4, 6,8,10,12,14,16,18, 20, 50, 100]

resultados_chebyshev_extra = []
for m in valores_m:
    for n in valores_n:
        d_max = min(m, n) - 1
        for d in valores_d:
            if d <= d_max:
                try:
                    indices = np.arange(1, m + 1)
                    X = np.cos(
                        (2 * indices - 1)
                        / (2 * m)
                        * np.pi
                    )
                    Y = np.linspace(-1.5, 1.5, n)

                    # Base canônica
                    LX_can = matriz_vander(X, d)
                    LY_can = matriz_vander(Y, d)
                    RX_can = np.linalg.pinv(LX_can)
                    T_can = LY_can @ RX_can
                    dados_can = estatisticas_svd(T_can)

                    # Base de Lagrange
                    k = np.arange(d + 1)
                    X_base = np.cos(
                        (2 * k + 1)
                        * np.pi
                        / (2 * (d + 1))
                    )
                    LX_lag = matriz_lagrange(X, X_base)
                    LY_lag = matriz_lagrange(Y, X_base)
                    RX_lag = np.linalg.pinv(LX_lag)
                    T_lag = LY_lag @ RX_lag
                    dados_lag = estatisticas_svd(T_lag)

                    # Comparações
                    S_can = np.linalg.svd(
                        T_can,
                        compute_uv=False
                    )
                    S_lag = np.linalg.svd(
                        T_lag,
                        compute_uv=False
                    )
                    erro_fro = np.linalg.norm(
                        T_can - T_lag,
                        ord="fro"
                    )
                    erro_max = np.max(
                        np.abs(T_can - T_lag)
                    )
                    resultados_chebyshev_extra.append({
                        "m": m,
                        "n": n,
                        "d": d,
                        "cond_LX_can":
                            np.linalg.cond(LX_can),
                        "cond_LX_lag":
                            np.linalg.cond(LX_lag),
                        "sigma_max_can":
                            dados_can["sigma_max"],
                        "sigma_min_can":
                            dados_can["sigma_min_nz"],
                        "cond_T_can":
                            dados_can["cond_T"],
                        "sigma_max_lag":
                            dados_lag["sigma_max"],
                        "sigma_min_lag":
                            dados_lag["sigma_min_nz"],
                        "cond_T_lag":
                            dados_lag["cond_T"],
                        "erro_fro":
                            erro_fro,
                        "erro_max":
                            erro_max,
                        "norma_T_can":
                            np.linalg.norm(T_can, 2),
                        "norma_T_lag":
                            np.linalg.norm(T_lag, 2),
                        "diferenca_svd":
                            np.linalg.norm(
                                S_can[:d+1]
                                - S_lag[:d+1]
                            )
                    })

                except Exception as e:
                    resultados_chebyshev_extra.append({
                        "m": m,
                        "n": n,
                        "d": d,
                        "cond_LX_can": np.nan,
                        "cond_LX_lag": np.nan,
                        "sigma_max_can": np.nan,
                        "sigma_min_can": np.nan,
                        "cond_T_can": np.nan,
                        "sigma_max_lag": np.nan,
                        "sigma_min_lag": np.nan,
                        "cond_T_lag": np.nan,
                        "erro_fro": np.nan,
                        "erro_max": str(e),

                    })

df_chebyshev_extra = pd.DataFrame(resultados_chebyshev_extra)
pd.set_option("display.float_format", "{:.3e}".format)
df_chebyshev_extra
```

### Experimento de predição

Neste experimento foi escolhida a função
$f(x)=\sin(5x)$
uma função suave e oscilatória, adequada para observar o comportamento de aproximações polinomiais. A partir de um conjunto de m=11 amostras, constrói-se um polinômio de grau d=10, tanto utilizando nós equiespaçados quanto nós de Chebyshev.

O objetivo é reproduzir o cenário clássico de interpolação: a partir dos valores da função em um conjunto de pontos \(X\subset[-1,1]\), calcula-se o polinômio interpolador associado e, em seguida, avalia-se esse polinômio em um conjunto maior de pontos

$Y=[-1.5,\,1.5]$

Dessa forma, parte das avaliações ocorre dentro do intervalo de treinamento $[-1,1]$, enquanto outra parte ocorre fora dele. Essa região externa corresponde a um problema de **extrapolação**, no qual o polinômio tenta prever valores da função em pontos para os quais nenhuma informação foi fornecida durante a construção do modelo.

O interesse do experimento é analisar até que ponto a interpolação produz boas aproximações fora do domínio observado e comparar o comportamento dos nós equiespaçados e dos nós de Chebyshev quando submetidos a esse processo de extrapolação.

```python

def funcao_alvo(x):
    return np.sin(5 * x)

m = 11 
d = 10 

X_equi = np.linspace(-1, 1, m)

indices = np.arange(1, m + 1)
X_cheb = np.cos((2 * indices - 1) / (2 * m) * np.pi)
Y_total = np.linspace(-1.5, 1.5, 500)



b_equi = funcao_alvo(X_equi)
b_cheb = funcao_alvo(X_cheb)
Y_verdadeiro = funcao_alvo(Y_total)

# CANONICA
X_base_equi = X_equi[:d+1]
Lx_equi = matriz_lagrange(
    X_equi,
    X_base_equi
)
Rx_equi = np.linalg.pinv(Lx_equi)
Ly_equi = matriz_lagrange(
    Y_total,
    X_base_equi
)
T_equi = Ly_equi @ Rx_equi
P_equi = T_equi @ b_equi


#LAGRANGE
X_base_cheb = X_cheb[:d+1]
Lx_cheb = matriz_lagrange(
    X_cheb,
    X_base_cheb
)
Rx_cheb = np.linalg.pinv(Lx_cheb)
Ly_cheb = matriz_lagrange(
    Y_total,
    X_base_cheb
)
T_cheb = Ly_cheb @ Rx_cheb
P_cheb = T_cheb @ b_cheb


plt.figure(figsize=(12, 7))
plt.axvspan(-1, 1, color='lightgreen', alpha=0.3, label='Domínio de Treino $X \in [-1, 1]$')
plt.axvspan(-1.5, 1, color='salmon', alpha=0.3, label='Domínio de Extrapolação $Y \in [-1.5, -1)')
plt.axvspan(1, 1.5, color='salmon', alpha=0.3, label='Domínio de Extrapolação $Y \in [1, 1.5)')
plt.plot(Y_total, Y_verdadeiro, 'k--', linewidth=2, label='Função Real $\sin(5x)$')
plt.plot(Y_total, P_equi, 'r-', linewidth=2, label='Polinômio (Equiespaçados em $[0,1]$)')
plt.plot(Y_total, P_cheb, 'b-', linewidth=2, label='Polinômio (Chebyshev em $[0,1]$)')
plt.ylim(-10, 10)
plt.xlim(-1.5, 1.5)
plt.title('Extrapolação em Domínios Disjuntos \nA Aceleração Violenta de Chebyshev Fora do Domínio', fontsize=14)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend(loc='upper left')
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Análise dos resultados

O gráfico mostra que o polinômio interpolador reproduz com excelente precisão a função $f(x)=\sin(5x)$ dentro do intervalo de treinamento $[-1,1]$. Nesse domínio, tanto os nós equiespaçados quanto os nós de Chebyshev produzem aproximações praticamente indistinguíveis da função original.

Entretanto, ao ultrapassar os limites do conjunto de treinamento, o comportamento muda drasticamente. Nas regiões de extrapolação $[-1.5,-1)$ e $(1,1.5]$, os polinômios passam a crescer rapidamente em módulo, afastando-se da função real. Esse fenômeno ocorre porque o polinômio foi ajustado apenas com informações contidas em $[-1,1]$, não havendo qualquer garantia de que seu comportamento permaneça adequado fora desse intervalo.

Observa-se ainda que os polinômios construídos com nós de Chebyshev e com nós equiespaçados apresentam comportamentos bastante semelhantes dentro do domínio de treinamento, mas diferem na extrapolação. Embora os nós de Chebyshev sejam conhecidos por reduzir oscilações e melhorar a estabilidade da interpolação em $[-1,1]$, essa vantagem desaparece fora do intervalo de interpolação. De fato, os polinômios associados aos nós de Chebyshev podem crescer muito rapidamente fora do domínio, comportamento relacionado ao crescimento dos próprios polinômios de Chebyshev para $|x|>1$.

O experimento evidencia uma distinção importante entre **interpolação** e **extrapolação**: uma aproximação pode ser extremamente precisa no intervalo onde foi construída e, ainda assim, produzir previsões completamente inadequadas logo após sair da região observada.

Experimento análogo ao item g). Dessa formaf, foi escolhida a função
$f(x)=\sin(5x)$
uma função suave e oscilatória, adequada para observar o comportamento de aproximações polinomiais. A partir de um conjunto de m=11 amostras, constrói-se um polinômio de grau d=10, tanto utilizando nós equiespaçados quanto nós de Chebyshev.

O objetivo é reproduzir o cenário clássico de interpolação: a partir dos valores da função em um conjunto de pontos \(X\subset[0,1]\), calcula-se o polinômio interpolador associado e, em seguida, avalia-se esse polinômio em um conjunto cujo único ponto em comum é o 0:

$Y=[-1,0]$

Dessa forma, as avaliações ocorrem totalmente fora  do intervalo de treinamento $[0,1]$. Essa região externa corresponde a um problema de **extrapolação**, no qual o polinômio tenta prever valores da função em pontos para os quais nenhuma informação foi fornecida durante a construção do modelo.

```python

def funcao_alvo(x):
    return np.sin(5 * x)

m = 11  
d = 10

X_equi = np.linspace(0, 1, m)
indices = np.arange(1, m + 1)
X_cheb_original = np.cos((2 * indices - 1) / (2 * m) * np.pi)
X_cheb_deslocado = (X_cheb_original + 1.0) / 2.0

Y_total = np.linspace(-1, 1, 500)

b_equi = funcao_alvo(X_equi)
b_cheb = funcao_alvo(X_cheb_deslocado)
Y_verdadeiro = funcao_alvo(Y_total)
X_base_equi = X_equi[:d+1]
Lx_equi = matriz_lagrange(X_equi, X_base_equi)
Rx_equi = np.linalg.pinv(Lx_equi) 
Ly_equi = matriz_lagrange(Y_total, X_base_equi)
T_equi = Ly_equi @ Rx_equi
P_equi = T_equi @ b_equi

X_base_cheb = X_cheb_deslocado[:d+1]
Lx_cheb = matriz_lagrange(X_cheb_deslocado, X_base_cheb)
Rx_cheb = np.linalg.pinv(Lx_cheb)

Ly_cheb = matriz_lagrange(Y_total, X_base_cheb)
T_cheb = Ly_cheb @ Rx_cheb
P_cheb = T_cheb @ b_cheb


plt.figure(figsize=(12, 7))
plt.axvspan(0, 1, color='lightgreen', alpha=0.3, label='Domínio de Treino $X \in [0, 1]$')
plt.axvspan(-1, 0, color='salmon', alpha=0.3, label='Domínio de Extrapolação $Y \in [-1, 0)$')
plt.plot(Y_total, Y_verdadeiro, 'k--', linewidth=2, label='Função Real $\sin(5x)$')
plt.plot(Y_total, P_equi, 'r-', linewidth=2, label='Polinômio (Equiespaçados em $[0,1]$)')
plt.plot(Y_total, P_cheb, 'b-', linewidth=2, label='Polinômio (Chebyshev em $[0,1]$)')
plt.ylim(-10, 10)
plt.xlim(-1, 1)
plt.title('Extrapolação em Domínios Disjuntos \nA Aceleração Violenta de Chebyshev Fora do Domínio', fontsize=14)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend(loc='upper left')
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Análise dos resultados

Os resultados obtidos são semelhantes aos observados no experimento anterior. No intervalo de treinamento $[0,1]$, os polinômios interpoladores reproduzem com alta precisão a função $f(x)=\sin(5)$, mostrando que a interpolação foi capaz de capturar adequadamente o comportamento da função nos pontos amostrados.

Entretanto, ao avaliar os polinômios na região $[-1,0)$, que não contém nenhum ponto de treinamento, observa-se uma deterioração significativa da aproximação. À medida que nos afastamos do domínio onde os dados foram coletados, os polinômios passam a divergir da função real e apresentam crescimento acentuado, caracterizando um processo de extrapolação instável.

O experimento reforça que a qualidade de uma aproximação polinomial está fortemente associada ao domínio em que ela foi construída. Mesmo quando a interpolação é praticamente perfeita nos dados de treinamento, não há garantia de que o modelo produza previsões confiáveis em regiões onde nenhuma informação foi observada.
