## Quetão 1a)


**Enunciado**: 
Mostre que, para
$Q_v x = \|x\| e_1$,
temos
$v = x - \|x\| e_1$
e
$\beta = \frac{1}{\|x\|(\|x\| - x_1)}$ e, equivalentemente,
$\beta=\frac{\|x\| + x_1}{\|x\|(\|x\|^2 - x_1^2)}=\frac{\|x\| + x_1}{\|x\| \cdot \|y\|^2}$
onde
$y = (x_2, \ldots, x_n)$.

### Parte 1 — Provar as igualdades para $\beta$

Desejamos mostrar que

$\beta=\frac{1}{\|x\|(\|x\|-x_1)}=\frac{\|x\|+x_1}{\|x\|(\|x\|^2-x_1^2)}=\frac{\|x\|+x_1}{\|x\|\|y\|^2}$,

onde

$y=(x_2,\ldots,x_n)$.

Começando por

$\beta=\frac{1}{\|x\|(\|x\|-x_1)}$,

multiplicamos numerador e denominador por $(\|x\|+x_1)$:

$\beta=\frac{\|x\|+x_1}{\|x\|(\|x\|-x_1)(\|x\|+x_1)}$.

Aplicando a diferença de quadrados,

$(\|x\|-x_1)(\|x\|+x_1)=
\|x\|^2-x_1^2$,

obtemos

$\beta=\frac{\|x\|+x_1}{\|x\|(\|x\|^2-x_1^2)}$.

Agora, como

$\|x\|^2=x_1^2+\|y\|^2$,

segue que

$\|x\|^2-x_1^2=\|y\|^2$.

Portanto,


$\boxed{\beta=\frac{\|x\|+x_1}{\|x\|\|y\|^2}}$.

Assim, todas as expressões dadas para $\beta$ são equivalentes.

### Parte 2 — Provar os valores de $v$ e $\beta$

Queremos mostrar que, tomando

$v=x-\|x\|e_1$

e

$\beta=\frac{1}{\|x\|(\|x\|-x_1)}$,

temos

$Q_vx=\|x\|e_1$,

onde

$Q_v=I-\beta vv^*$.

Substituindo $v=x-\|x\|e_1$,

$Q_v=I-\frac{1}{\|x\|(\|x\|-x_1)}(x-\|x\|e_1)(x-\|x\|e_1)^*$.

Expandindo o produto,

$(x-\|x\|e_1)(x-\|x\|e_1)^*=xx^*-\|x\|xe_1^*-\|x\|e_1x^*+\|x\|^2e_1e_1^*$.

Logo,

$Q_vx=\left(I-\frac{xx^*-\|x\|xe_1^*-\|x\|e_1x^*+\|x\|^2e_1e_1^*}{\|x\|(\|x\|-x_1)}\right)x$.

Aplicando a matriz ao vetor $x$,

$Q_vx=x-\frac{xx^*x-\|x\|xe_1^*x-\|x\|e_1x^*x+\|x\|^2e_1e_1^*x}{\|x\|(\|x\|-x_1)}$.

Usamos agora as identidades

$x^*x=\|x\|^2$,

$e_1^*x=x_1$,

$e_1e_1^*x=x_1e_1$.

Substituindo,

$Q_vx=x-\frac{\|x\|^2x-\|x\|x_1x-\|x\|^2e_1+\|x\|^2x_1e_1}{\|x\|(\|x\|-x_1)}$.

Agrupando os termos com $x$,

$Q_vx=x-\frac{\|x\|(\|x\|-x_1)x-\|x\|^2(\|x\|-x_1)e_1}{\|x\|(\|x\|-x_1)}$.

Cancelando o fator comum $\|x\|(\|x\|-x_1)$,

$Q_vx=x-(x-\|x\|e_1)$.

Finalmente,

$\boxed{Q_vx=\|x\|e_1}$

Portanto, os valores

$\boxed{v=x-\|x\|e_1}\qquad\text{e}\qquad\boxed{\beta=\frac{1}{\|x\|(\|x\|-x_1)}}$

produzem exatamente o refletor desejado.

## Questão 1b)

**Enunciado:** Verifique que, dependendo do sinal de $x_1$, uma das fórmulas acima para $\beta$ não está
sujeita a perda de precisão devido a cancelamento numérico.

### Estabilidade Numérica das Fórmulas para $\beta$

Considere as expressões equivalentes para $\beta$:

$\beta=\frac{1}{\|x\|(\|x\|-x_1)}=\frac{\|x\|+x_1}{\|x\|(\|x\|^2-x_1^2)}=\frac{\|x\|+x_1}{\|x\|\|y\|^2}$,

onde

$y=(x_2,\ldots,x_n)$.

### Caso $x_1 > 0$

O erro de cancelamento ocorre quando

$\|x\| \approx x_1$.

Nesse caso, a primeira fórmula contém o termo

$\|x\|-x_1$,

isto é, a subtração entre dois números muito próximos. Em aritmética de ponto flutuante, essa operação pode provocar perda significativa de precisão.

Assim,

$\beta=\frac{1}{\|x\|(\|x\|-x_1)}=\frac{\|x\|+x_1}{\|x\|(\|x\|^2-x_1^2)}$

é numericamente instável quando

$\|x\| \approx x_1$.

Utilizando a identidade:

$\beta=\frac{\|x\|+x_1}{\|x\|\|y\|^2}$

não aparece a diferença $\|x\|-x_1$ e $\|x\|^2-x_1^2$, que é justamente a responsável pelo cancelamento. Portanto, quando

$\|x\| \approx x_1$,

a expressão

$\boxed{\beta=\frac{\|x\|+x_1}{\|x\|\|y\|^2}}$

é preferível do ponto de vista numérico.

### Caso $x_1 < 0$

Agora o termo problemático deixa de ser

$\|x\|-x_1$,

pois

$\|x\|-x_1=\|x\|+|x_1|$,

ou seja, trata-se de uma soma de números positivos. Não há cancelamento nessa operação.

Consequentemente,

$\boxed{\beta=\frac{1}{\|x\|(\|x\|-x_1)}}$

é numericamente estável quando $x_1<0$.

Por outro lado, a forma racionalizada contém

$\|x\|+x_1$.

Se

$\|x\|\approx -x_1$,

então $\|x\|+x_1 \approx 0$ é a diferença entre números quase iguais, surgindo novamente o erro de cancelamento.

Logo, para $x_1<0$, a expressão

$\beta=\frac{\|x\|+x_1}{\|x\|\|y\|^2}$

é a que pode sofrer perda de precisão.


### Conclusão

- Se $x_1>0$, ocorre cancelamento em $\|x\|-x_1$. Nesse caso, deve-se utilizar

$\boxed{\beta=\frac{\|x\|+x_1}{\|x\|\|y\|^2}}$

- Se $x_1<0$, ocorre cancelamento em $\|x\|+x_1$. Nesse caso, deve-se utilizar

$\boxed{\beta=\frac{1}{\|x\|(\|x\|-x_1)}}$

Em resumo, a forma mais estável é aquela que evita a subtração entre números de mesma magnitude e sinais opostos.

## Questão 1c)

**Enunciado:** Escreva uma função reflector $(x)$ que, dado um vetor x, calcula o vetor v e o escalar
$\beta$ do refletor de Householder Q tal que $Q_x = ∥x∥e1$.

````python
import numpy as np

def refletor(x):
    xn = np.linalg.norm(x).astype(x.dtype)
    if xn == 0:
        #usa o tipo exato do vetor
        return np.eye(x.size, dtype=x.dtype)
    
    # herda o tipo do vetor original
    e1 = np.zeros(x.size, dtype=x.dtype)
    e1[0] = 1  # usar o int em vez do float

    v = x - xn*e1
    I = np.eye(x.size, dtype=x.dtype)
    y = x[1:]

    if x[0] > 0:
        yn = np.square(np.linalg.norm(y).astype(y.dtype))
        if yn == 0:
            beta = 0  # usar o int em vez do float
        else:
            beta = (xn +x[0])/(xn *yn)
    else:
        beta = 1/(xn *(xn-x[0]))
    
    return v, beta
````

## Questão 1d)

**Enunciado:** Condicionamento. Mostre que a derivada de v em relação a x é $J = I − e_1x^{*}/∥x∥$.
Deduza que o número de condicionamento absoluto de v em relação a x é menor ou igual
a 2.

### Derivada de $v$ em relação a $x$

Queremos mostrar que a derivada de

$v=x-\|x\|e_1$

em relação a $x$ é

$J=I-\frac{e_1x^*}{\|x\|}$.

Denotemos por $J$ a matriz Jacobiana de $v$ em relação a $x$:

$J=\begin{bmatrix}\dfrac{\partial v_1}{\partial x_1} & \cdots & \dfrac{\partial v_1}{\partial x_n}\\\vdots & \ddots & \vdots\\\dfrac{\partial v_n}{\partial x_1} & \cdots & \dfrac{\partial v_n}{\partial x_n}\end{bmatrix}$.

Vamos analisar as componentes de $v$.

#### Primeira componente

Temos

$v_1=x_1-\|x\|=x_1-\sqrt{x_1^2+\cdots+x_n^2}$.

Logo,

$\frac{\partial v_1}{\partial x_1}=1-\frac{2x_1}{2\|x\|}=1-\frac{x_1}{\|x\|}$.

Para $j\neq 1$,

$\frac{\partial v_1}{\partial x_j}=-\frac{x_j}{\|x\|}$

Portanto, a primeira linha da Jacobiana é

$\left(1-\frac{x_1}{\|x\|},-\frac{x_2}{\|x\|},\ldots,-\frac{x_n}{\|x\|}\right)$.

#### Demais componentes

Para $i>1$,

$v_i=x_i$.

Assim,

$\frac{\partial v_i}{\partial x_j}=\begin{cases}1, & i=j,\\[2mm]0, & i\neq j.\end{cases}$

Logo, todas as demais linhas coincidem com as da identidade.

Portanto,

$J=\begin{bmatrix}1-\dfrac{x_1}{\|x\|}&-\dfrac{x_2}{\|x\|}&\cdots&-\dfrac{x_n}{\|x\|}\\[3mm]0&1&\cdots&0\\\vdots&&\ddots&\vdots\\0&0&\cdots&1\end{bmatrix}$.

Agora observemos que

$e_1x^*=\begin{bmatrix}x_1 & x_2 & \cdots & x_n\\0 & 0 & \cdots & 0\\\vdots & \vdots & \ddots & \vdots\\0 & 0 & \cdots & 0\end{bmatrix}$.


Dividindo por $\|x\|$,

$\frac{e_1x^*}{\|x\|}=\begin{bmatrix}\dfrac{x_1}{\|x\|}&\dfrac{x_2}{\|x\|}
&\cdots
&\dfrac{x_n}{\|x\|}\\0&0&\cdots&0\\\vdots&\vdots&\ddots&\vdots\\0&0&\cdots&0\end{bmatrix}$.

Portanto,

$I-\frac{e_1x^*}{\|x\|}=\begin{bmatrix}1-\dfrac{x_1}{\|x\|}&-\dfrac{x_2}{\|x\|}&\cdots&-\dfrac{x_n}{\|x\|}\\0&1&\cdots&0\\\vdots&&\ddots&\vdots\\0&0&\cdots&1\end{bmatrix}$,

que coincide exatamente com a Jacobiana calculada.

Assim,

$\boxed{J=I-\frac{e_1x^*}{\|x\|}}$.


### Número de condicionamento absoluto de \(v\)

Da parte anterior, sabemos que a Jacobiana de

$v=x-\|x\|e_1$

é

$J=I-\frac{e_1x^*}{\|x\|}$.

Definindo

$u=\frac{x}{\|x\|}$,

podemos escrever

$J=I-e_1u^*$.

Observe que

$\|u\|=1$,

isto é, $u$ é um vetor unitário.

O número de condicionamento absoluto de $v$ em relação a $x$ é dado por

$\kappa_{\mathrm{abs}}(v,x)=\|J\|$.

Portanto, basta mostrar que

$\|J\|\leq 2$.

Pela definição da norma matricial induzida,

$\|J\|=\max_{\|y\|=1}\|Jy\|$.

Seja então $y$ um vetor unitário arbitrário. Temos:

$Jy=(I-e_1u^*)y=y-e_1(u^*y)$.

Aplicando a desigualdade triangular,

$\|Jy\|=\|y-e_1(u^*y)\|\leq\|y\|+\|e_1(u^*y)\|$.

Como $\|y\|=1$ e $\|e_1\|=1$,

$\|e_1(u^*y)\|=|u^*y|$.

Além disso, pela desigualdade de Cauchy--Schwarz,

$|u^*y|\leq\|u\|\,\|y\|=1$.

Logo,

$\|Jy\|\leq1+1=2$.

Como essa desigualdade vale para todo vetor unitário $y$,

$\|J\|=\max_{\|y\|=1}\|Jy\|\leq2$.

Portanto,

$\boxed{\kappa_{\mathrm{abs}}(v,x)=\|J\|\leq2}$

Assim, o número de condicionamento absoluto de $v$ em relação a $x$ é menor ou igual a 2.

Primeiramente, a função refletor será testada em um exemplo simples. O objetivo é apenas mostrar que ela funciona e mantém os tipos do input, ou seja, se o input é float64, o output é float64.

````python
def calcular_erro(v_32, v_64, b_32, b_64):
    erro_v= np.linalg.norm(v_32-v_64)/np.linalg.norm(v_64)
    erro_b = np.abs((b_32-b_64)/b_64)
    return erro_v, erro_b
````

Para que o refletor exista e produza uma reflexão não-trivial, o vetor $v$ não pode ser nulo ($vv^T \neq 0$) e $\beta \neq 0$. Portanto, para que a equação seja verdadeira, o termo entre parênteses deve ser igual a zero:

**Enunciado:** (Bônus) Mostre/Verifique que $v_1$ é (relativamente!) mal-condicionado quando x está
quase alinhado com $e_1$. Inspirado na fórmula para $\beta$, proponha outra fórmula para calcular
$v_1$ (ainda refletindo x em $∥x∥e_1$) que não seja tão sensível a $x_1$.
Obs: O objetivo de refletir um vetor com $x_1 > 0$ em um vetor com a primeira coordenada
positiva é garantir que a matriz R terá elementos positivos na diagonal.

A definição original da primeira coordenada do vetor refletor é:$$v_1 = x_1 - \|x\|$$
Se o vetor $x$ está quase alinhado ao eixo $e_1$ e $x_1 > 0$, as outras coordenadas são pequenas. Isso significa que a norma do vetor é praticamente igual à sua primeira coordenada ($\|x\| \approx x_1$).Ao calcularmos $v_1$, teremos $v_1 \approx x_1 - x_1 \approx 0$. O computador é forçado a subtrair dois números positivos extremamente próximos, o que causa uma perda de precisão e amplificação do erro numérico. Para gerar uma fórmula imune a esse mal-condicionamento, multiplicamos e dividimos a equação original pelo seu conjugado, $(x_1 + \|x\|)$:

$$v_1 = \frac{(x_1 - \|x\|)(x_1 + \|x\|)}{x_1 + \|x\|}$$

Aplicando o produto da soma pela diferença no numerador:

$$v_1 = \frac{x_1^2 - \|x\|^2}{x_1 + \|x\|}$$

Sabemos que a norma ao quadrado de um vetor é a soma do quadrado de todos os seus elementos. Podemos separar o primeiro elemento ($x_1$) do resto do vetor (que chamaremos de $y$):

$$\|x\|^2 = x_1^2 + \|y\|^2$$

Substituindo isso na nossa equação:

$$v_1 = \frac{x_1^2 - (x_1^2 + \|y\|^2)}{x_1 + \|x\|}$$

Os termos $x_1^2$ se cancelam, resultando na fórmula final:

$$v_1 = \frac{-\|y\|^2}{x_1 + \|x\|}$$

Essa fórmula é melhor por quê o numerador ($-\|y\|^2$) é uma soma de quadrados e o denominador ($x_1 + \|x\|$) soma dois números estritamente positivos. Sem subtrações de termos semelhantes, a fórmula torna-se completamente livre de instabilidades numéricas.
