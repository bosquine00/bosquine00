## Forma de Hessemberg
Se você se sentir à vontade, pode definir um tipo de dado para os refletores, e então as funções
desta questão poderiam receber e retornar Refletores, ao invés de v e β separadamente.

## Questão 2a)

**Enunciado:**
Aplicando refletores. Escreva uma função apply_reflector($v,\beta, b$) que calcula $Qvb$, onde $Qv = I −\beta vv∗$ é o refletor de Householder dado por v e $\beta$.

````python
import numpy as np

def apply_reflector(v, beta, b):
    produto_escalar = np.dot(v, b)
    Qb = b - (beta* produto_escalar)*v

    return Qb
````

### Complexidade da Função `apply_reflector_adaptada`

Considere a função

Desejamos determinar sua complexidade assintótica em função de $n$, onde $n$ representa o tamanho do vetor $v$.

#### Passo 1: Cópia do vetor

A instrução

```python
b_out = np.copy(b)
```

realiza uma cópia completa de $b$. Se considerarmos apenas a parte do algoritmo que aplica o refletor, essa operação não depende de $v$. Em aplicações típicas do algoritmo de Hessenberg, o tamanho de $b$ é proporcional ao tamanho do subproblema considerado.

#### Passo 2: Extração do trecho relevante

```python
b_resto = b_out[(len(b)-len(v)):]
```

A operação de fatiamento produz uma visão ou cópia de um trecho contendo exatamente $n$ elementos. Portanto, o custo é proporcional a $n$: $O(n)$.

#### Passo 3: Produto escalar

```python
produto_escalar = np.dot(v, b_resto)
```

O produto escalar entre dois vetores de tamanho $n$ exige:

- $n$ multiplicações;
- $n-1$ somas.

Logo, seu custo é $O(n)$.

#### Passo 4: Atualização do vetor

```python
b_resto - (beta * produto_escalar) * v
```

A expressão realiza:

1. uma multiplicação escalar de $v$ por um número;
2. uma subtração elemento a elemento.

Ambas percorrem todos os $n$ elementos do vetor, resultando em O(n)$.

A atribuição ao trecho correspondente de `b_out`

```python
b_out[(len(b)-len(v)):] = ...
```

também percorre $n$ posições, mantendo o mesmo custo assintótico.

### Complexidade Total

Somando os custos dominantes: $O(n) + O(n) + O(n) + O(n)$.

Como constantes multiplicativas são desprezadas na notação assintótica, obtemos $O(4n)=O(n)$.

## Questão 2c)

**Enunciado:** Generalize sua função para apply_reflector($v, \beta, A$) que calcula $QvA$ para uma
matriz A (com no mínimo o mesmo número de linhas do que v).

````python
def apply_reflector_c(v, beta, A):
    # QA = (I - Bvv*)A = A- Bvv*A
    A_out = np.copy(A)
    m = len(v)
    k = A.shape[0]-m

    A_resto = A_out[k:, :]
    va = np.dot(v, A_resto)
    A_out[k:,:] = A_resto - beta * np.outer(v,va)

    return A_out
````

## Questão 2d)

**Enunciado:** Escreva uma função rev_apply_reflector($v, \beta, A$) que calcula $AQv$ (com A de
dimensões compatíveis).

````python
def rev_apply_reflector(v, beta, A):
    # AQ = A(I - beta*vv*) = A - beta*(Av)v*
    A_out = np.copy(A)
    
    M, N = A.shape
    m = len(v)
    k = N - m
    A_resto = A_out[:, k:]
    
    # y = A_resto @ v
    y = A_resto @ v
    A_out[:, k:] = A_resto - beta * np.outer(y, v)
    
    return A_out
````

## Questão 2e)

**Enunciado:** Juntando tudo. Escreva uma função to_hessemberg(A) que calcula a forma de Hessemberg de uma matriz A usando refletores de Householder. A função deve retornar uma
lista de refletores ($v_i,\beta_i$), a matriz H tal que $A = QHQ∗$, e, opcionalmente, Q, que é a
matriz ortogonal dada pelo produto dos refletores.

````python
def to_hessemberg(A):
    linhas = A.shape[0]
    H = np.copy(A)
    
    #matriz Q começa como Identidade e vai acumulando as transformações
    Q = np.identity(linhas)
    refletores=[]

    # iteramos até n-2, pois as duas últimas linhas formam um bloco 2x2 que já estão na forma de Hessenberg.
    for k in range(linhas - 2):
        x = H[k+1:, k]
        n_x = np.linalg.norm(x)
        
        if n_x <1e-12:
            continue
        
        #cálculo de v e beta
        v = np.copy(x)
        if x[0] > 0:
            y = x[1:]
            v[0] = -np.sum(np.square(y)) / (x[0] + n_x)
        else:
            v[0] = x[0] - n_x
        beta = 2.0 / np.dot(v, v)
        refletores.append((v, beta))
        
        #aplicando o refletor pela esquerda(linhas)
        H[:, k:] = apply_reflector_c(v, beta, H[:, k:])
        
        #aplicando o refletor pela direit(colunas)
        H = rev_apply_reflector(v, beta, H)

        Q = rev_apply_reflector(v, beta, Q)
    H = np.triu(H, -1)
    return refletores, H, Q
````

````python
#APENAS UM TESTE BÁSICO
A = np.array([
        [ 2.0,  1.0,  3.0,  4.0],
        [ 5.0,  6.0,  1.0,  2.0],
        [ 9.0, -2.0,  4.0,  1.0],
        [-1.0,  3.0,  5.0,  8.0]
    ])
    
print("MATRIZ A:")
print(A)
refletores, H, Q = to_hessemberg(A)

print("\n Hessenberg:")
print(np.round(H, 4))
````

## Questão 2f)

**Enunciado:** Verifique que sua função de fato está correta, calculando $∥A − QHQ∗∥$ e $∥Q∗Q − I∥$ para
matrizes simétricas e não simétricas, e de tamanhos 2, 10 e 100.

````python
import pandas as pd

resultados = []
tamanhos = [2, 10, 100]

for n in tamanhos:
    I_n = np.identity(n)

    # MATRIZ NÃO SIMÉTRICA
    A_nao_simetrica = np.random.randn(n, n)
    _, H_ns, Q_ns = to_hessemberg(A_nao_simetrica)

    erro_ns = np.linalg.norm(A_nao_simetrica - Q_ns @ H_ns @ Q_ns.T)
    erro_orto_ns = np.linalg.norm(Q_ns.T @ Q_ns - I_n)

    resultados.append({
        "Dimensão": n,
        "Tipo": "Não Simétrica",
        "||A - QHQ*||": erro_ns,
        "||Q*Q - I||": erro_orto_ns
    })

    # MATRIZ SIMÉTRICA
    M = np.random.randn(n, n)
    A_simetrica = M + M.T

    _, H_s, Q_s = to_hessemberg(A_simetrica)

    erro_s = np.linalg.norm(A_simetrica - Q_s @ H_s @ Q_s.T)
    erro_orto_s = np.linalg.norm(Q_s.T @ Q_s - I_n)

    resultados.append({
        "Dimensão": n,
        "Tipo": "Simétrica",
        "||A - QHQ*||": erro_s,
        "||Q*Q - I||": erro_orto_s
    })

df_resultados = pd.DataFrame(resultados)

# Exibição em notação científica
pd.set_option("display.float_format", "{:.2e}".format)

display(df_resultados)
````

Para as matrizes de dimensões $10\times10$ e $100\times100$, os erros de reconstrução permaneceram entre $10^{-15}$ e $10^{-13}$, enquanto os erros de ortogonalidade ficaram próximos de $10^{-15}$. Ainda assim, os erros permanecem extremamente pequenos quando comparados à escala dos dados.

Por outro lado, a ortogonalidade da matriz $Q$ é preservada com alta precisão em todos os experimentos, apresentando erros da ordem de $10^{-15}$. Isso indica que o produto sucessivo dos refletores de Householder foi realizado de forma numericamente estável.
